#### **Machine Learning 1 Definitions Part A:**

| # | Term | Definition |
|---|------|------------|
| 1 | Probabilistic model | A type of model that deals with uncertainty by specifying the probabilities of different possible outcomes based on the given input variables. |
| 2 | Prediction function | The function used in a model to estimate the expected outcome based on the input variables and model parameters. |
| 3 | Well-specified and misspecified models | A model is considered well-specified if it accurately represents the underlying process generating the data; it is misspecified if it fails to capture the essential structure, leading to potential errors in predictions. |
| 4 | Likelihood function | Represents how likely it is that a given set of parameters of the model could result in the observed data. |
| 5 | Maximum likelihood estimate | The parameter values that maximise the likelihood function, indicating the most probable parameter set given the observed data. |
| 6 | Connection to training error minimisation | Describes how reducing the training error, or the discrepancies between the model predictions and the actual outcomes on the training data, aligns with maximising the likelihood function in certain statistical models. |

#### **Machine Learning 1 Definitions Part B:**

| # | Term | Definition |
|---|------|------------|
| 7 | Bootstrap sample | A dataset created by randomly selecting individuals from the original dataset with replacement, used to estimate statistical properties. |
| 8 | Epistemic uncertainty | The uncertainty due to lack of knowledge, which can potentially be reduced by obtaining more information or data. |
| 9 | Parameter uncertainty | Uncertainty about the accuracy or stability of the estimated parameters within a model. |
| 10 | Model uncertainty | The uncertainty about whether the chosen model is the correct one, reflecting doubts about the model's structure. |
| 11 | Aleatoric uncertainty | The type of uncertainty that comes from the inherent randomness or variability in the model or system, which cannot be reduced by additional data. |
| 12 | Bootstrap method | A technique that improves statistical accuracy by randomly resampling the dataset with replacement to create multiple simulated samples for estimation. |

#### **Machine Learning 2 Definitions Part C:**

| #  | Term                      | Definition |
|----|---------------------------|------------|
| 13 | Predictive distribution   | Summarises the range of possible outcomes for new observations based on the current model and all observed data, incorporating the known uncertainties in the predictions. |
| 14 | Out-of-bag                | Refers to the data points in the original dataset that are not included in a bootstrap sample and are often used as a test set to assess model accuracy. |
| 15 | Maximum posterior estimate | Refers to the values of parameters that are most likely after accounting for both the prior knowledge and the evidence from the observed data. |
| 16 | Prior distribution        | The assumed distribution of a model's parameters before any actual data is observed, based on prior beliefs or knowledge. |
| 17 | Marginal likelihood       | Represents the probability of observing the collected data under all possible values of the parameters, considering the prior beliefs about these parameters. |
| 18 | Posterior distribution    | The updated distribution of a model's parameters after observing data, combining prior beliefs with the evidence from the data. |
