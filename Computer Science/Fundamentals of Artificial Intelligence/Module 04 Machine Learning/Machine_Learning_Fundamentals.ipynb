### **Machine Learning**

Machine Learning is a branch of Artifiicial Intelligence that involves programs that **Learns From Data**. Machine Learning performs **Inductive Learning**, which is learning based on training data or examples. A Machine Learning program makes **Predictions** and **Learns from Feedback** on the correctness of its predicitions. It is used in a multitude of applications, not limited to speech recognition, game playing and recommendation systems.

### **Supervised Learning**

Supervised Learning is a subset of Machine Learning where a **set of observations (training data)** is **Accompanied by Labels** indicating the class of each data and the known target value of each observations. The Supervised Learning program task is learn a model from training data to **Classify or Predict** the target value of a new data. There are two main techniques in supervised learning, the first being **Classification** and the second being **Regression**.
