### **Machine Learning**

Machine Learning is a branch of Artifiicial Intelligence that involves programs that **Learns From Data**. Machine Learning performs **Inductive Learning**, which is learning based on training data or examples. A Machine Learning program makes **Predictions** and **Learns from Feedback** on the correctness of its predicitions. It is used in a multitude of applications, not limited to speech recognition, game playing and recommendation systems.

### **Supervised Learning**

Supervised Learning is a subset of Machine Learning where a **set of observations (training data)** is **Accompanied by Labels** indicating the class of each data and the known target value of each observations. The Supervised Learning program task is learn a model from training data to **Classify or Predict** the target value of a new data. There are two main techniques in supervised learning, the first being **Classification** and the second being **Regression**.

### **Classification: Performance and Confusion Matrix**

When assessing how well a classification model performs, three key metrics are commonly used: accuracy, recall, and precision. These metrics can be understood through the variables shown in the confusion matrix below. A Confusion Matrix is a Table that Summarizes the Performance of a Classification Algorithm.

**Confusion Matrix**

|                | **Actual C** | **Actual C'** |
|----------------|--------------|---------------|
| **Predicted C** | a           | b             |
| **Predicted C'** | c           | d             |

### **Evaluation Metrics**

**Accuracy** measures the overall correctness of the classifier by calculating what proportion of all predictions were correct.

$$\frac{a+d}{a+b+c+d} = \frac{\text{total correct predictions}}{\text{all predictions}}$$

**Recall** (also known as sensitivity) measures how well the classifier identifies positive cases by calculating what fraction of actual positive instances were correctly identified.

$$\frac{a}{a+c} = \frac{\text{true positives}}{\text{all actual positives}}$$

**Precision** measures the reliability of positive predictions by calculating what fraction of instances predicted as positive were actually positive.

$$\frac{a}{a+b} = \frac{\text{true positives}}{\text{all predicted positives}}$$

### **Overfitting and Underfitting**

The goal is to aim for the best classification results on the training data but if the focus was only on training data, the classifier might overfit the training data and perform poorly on unseen data. 

| Aspect | **Underfitting** | **Overfitting** | **Optimal Fitting** |
|--------|------------------|-----------------|---------------------|
| **Definition** | Model too simple to capture patterns | Model memorizes training data including noise | Model captures true patterns without noise |
| **Training Performance** | Poor | Excellent | Good |
| **Test Performance** | Poor | Poor | Good |
| **Bias-Variance** | High bias, low variance | Low bias, high variance | Balanced bias-variance |
| **Problem** | Misses important patterns | Memorizes specific examples | Generalizes well |
| **Signs** | • Low training accuracy<br>• Low test accuracy<br>• Large prediction errors | • High training accuracy<br>• Much lower test accuracy<br>• Large train-test gap | • Good training accuracy<br>• Similar test accuracy<br>• Small train-test gap |
| **Analogy** | Student too lazy to study properly | Student memorizes answers without understanding | Student understands concepts and applies them |
| **Solutions** | • Use more complex model<br>• Add more features<br>• Reduce regularization<br>• Train longer | • Use simpler model<br>• Add regularization<br>• Get more data<br>• Early stopping | • Monitor with validation<br>• Use cross-validation<br>• Apply appropriate regularization |

### **Overfitting: Definition and Causes**

Given a classifier model **M₁**, we say **M₁** overfits the data if there exists another model **M₂** such that:

- **M₂** has smaller test error than **M₁** on test examples
- **M₂** has larger training error than **M₁** on training set

This definition captures the essence of overfitting: a model that performs exceptionally well on training data but poorly on new data is considered overfitted if a simpler model (with worse training performance) actually generalizes better. The table below shows some of the common causes of overfitting.

| Cause | Explanation | Example |
|-------|-------------|---------|
| **Noisy Data** | Model learns from random fluctuations and outliers instead of true patterns | Training data contains measurement errors or mislabeled examples |
| **Insufficient Training Data** | Not enough examples for the model to learn generalizable patterns | Using 100 samples to train a model with 1000 parameters |
| **Overly Complex Model** | Model has too many parameters relative to the amount of training data | Using a deep neural network for a simple linear relationship |

### **Model Complexity**

More sophisticated models have the capacity to capture intricate patterns in data. However, there's a critical balance to strike:

**The Complexity Sweet Spot**: Initially, as we increase model complexity, performance improves because the model can better capture underlying patterns. However, beyond an optimal point, additional complexity becomes counterproductive. The model begins learning noise and random variations specific to the training data that don't represent true patterns in the broader data distribution. This leads to overfitting.

### **Features Machine Learning Methods**

| Feature | Description | Considerations |
|---------|-------------|----------------|
| **Performance Metrics** | Standards used to evaluate model effectiveness | • Accuracy, precision, recall vary by problem<br>• Different applications prioritize different metrics<br>• Choose metrics aligned with business objectives |
| **Statistical Significance** | Model must outperform baseline approaches | • Must beat naive/random models<br>• Improvements should be statistically meaningful<br>• Consider confidence intervals and p-values |
| **Interpretability** | Ability to understand model decisions | • Can humans trace how conclusions were reached?<br>• Important for regulated industries<br>• Trade-off between accuracy and explainability |
| **Robustness** | Model stability under imperfect conditions | • Handles missing data gracefully<br>• Resistant to outliers and noise<br>• Maintains performance with data quality issues |
| **Scalability** | Computational efficiency and growth handling | • Training time with current data<br>• Performance impact as data size increases<br>• Resource requirements (memory, processing power) |

### **Model Selection Strategy**

When choosing a machine learning method, evaluate each feature based on your specific use case:
- **High-stakes decisions**: Prioritize interpretability and robustness
- **Large-scale applications**: Focus on scalability and computational efficiency  
- **Regulatory environments**: Emphasize significance and interpretability
- **Real-time systems**: Consider scalability and robustness to data issues
