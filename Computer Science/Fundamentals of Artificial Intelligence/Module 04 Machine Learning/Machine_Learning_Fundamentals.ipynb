### **Machine Learning**

Machine Learning is a branch of Artifiicial Intelligence that involves programs that **Learns From Data**. Machine Learning performs **Inductive Learning**, which is learning based on training data or examples. A Machine Learning program makes **Predictions** and **Learns from Feedback** on the correctness of its predicitions. It is used in a multitude of applications, not limited to speech recognition, game playing and recommendation systems.

### **Supervised Learning**

Supervised Learning is a subset of Machine Learning where a **set of observations (training data)** is **Accompanied by Labels** indicating the class of each data and the known target value of each observations. The Supervised Learning program task is learn a model from training data to **Classify or Predict** the target value of a new data. There are two main techniques in supervised learning, the first being **Classification** and the second being **Regression**.

### **Classification: Performance and Confusion Matrix**

When assessing how well a classification model performs, three key metrics are commonly used: accuracy, recall, and precision. These metrics can be understood through the variables shown in the confusion matrix below. A Confusion Matrix is a Table that Summarizes the Performance of a Classification Algorithm.

**Confusion Matrix**

|                | **Actual C** | **Actual C'** |
|----------------|--------------|---------------|
| **Predicted C** | a           | b             |
| **Predicted C'** | c           | d             |

### **Evaluation Metrics**

**Accuracy** measures the overall correctness of the classifier by calculating what proportion of all predictions were correct.

$$\frac{a+d}{a+b+c+d} = \frac{\text{total correct predictions}}{\text{all predictions}}$$

**Recall** (also known as sensitivity) measures how well the classifier identifies positive cases by calculating what fraction of actual positive instances were correctly identified.

$$\frac{a}{a+c} = \frac{\text{true positives}}{\text{all actual positives}}$$

**Precision** measures the reliability of positive predictions by calculating what fraction of instances predicted as positive were actually positive.

$$\frac{a}{a+b} = \frac{\text{true positives}}{\text{all predicted positives}}$$
