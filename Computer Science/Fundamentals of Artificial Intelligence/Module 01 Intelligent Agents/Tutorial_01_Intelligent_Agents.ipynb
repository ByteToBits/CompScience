### 1) Backtrack

Specify a state representation, operators, and goal test to solve the N-queen problem with N=4:

Four queens have to be placed on a 4x4 board where no row will have more than one queen, no column will have more than one queen, and no diagonal will have more than one queen.
Illustrate the workings of the backtracking algorithm (Backtrack1) to solve this problem.


**State Representation**: A List of of tiles $[ 0 0 0 0 ]$ is used to represent that no questions are placed on the board, and each position in the list represents the **Row Number** of the Queens between 1 to 4. The illustration below shows the arrangment of the queens as $[ 3 1 2 4 ]$ for example, which means 
- Queen 1: column 1, row 3
- Queen 2: column 2, row 1
- Queen 3: column 3, row 2
- Queen 4: column 4, row 4



**Operators**
Assign the tile value for the next unassigned queen, i.e., Ai means "assign the queen to row i, and to the first empty column". So Ai can be applied at most 4 times.

**Goal Test Check Rows**: All the numbers in the list are different.

**Goal Test Check Column**: You don't need to do column check for this representation as no queen will be in the same column.

**Check Diagonals**: Any pair of queen should have different row and column difference. If a queen is in row r1, column c1 and another queen in r2 and c2, then |r1r2|  |c1c2| should hold for any pair of queens. For instance, if the board is $[ 3 1 2 4 ]$, the second queen (r1 = 1, c1 = 2) and the third queen (r2 = 2, c2 = 3) are attacking each other diagonally because |1  2| = |2  3|.
