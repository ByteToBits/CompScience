### **1) Backtrack**

Specify a state representation, operators, and goal test to solve the N-queen problem with N=4:

Four queens have to be placed on a 4x4 board where no row will have more than one queen, no column will have more than one queen, and no diagonal will have more than one queen.
Illustrate the workings of the backtracking algorithm (Backtrack1) to solve this problem.


**State Representation**: A List of of tiles $[ 0 0 0 0 ]$ is used to represent that no questions are placed on the board, and each position in the list represents the **Row Number** of the Queens between 1 to 4. The illustration below shows the arrangment of the queens as $[ 3 1 2 4 ]$ for example, which means 
- Queen 1: column 1, row 3
- Queen 2: column 2, row 1
- Queen 3: column 3, row 2
- Queen 4: column 4, row 4



**Operators**
Assign the tile value for the next unassigned queen, i.e., Ai means "assign the queen to row i, and to the first empty column". So Ai can be applied at most 4 times.

**Goal Test Check Rows**: All the numbers in the list are different.

**Goal Test Check Column**: You don't need to do column check for this representation as no queen will be in the same column.

**Check Diagonals**: Any pair of queen should have different row and column difference. If a queen is in row r1, column c1 and another queen in r2 and c2, then |r1r2|  |c1c2| should hold for any pair of queens. For instance, if the board is $[ 3 1 2 4 ]$, the second queen (r1 = 1, c1 = 2) and the third queen (r2 = 2, c2 = 3) are attacking each other diagonally because |1  2| = |2  3|.

### **4) Search**

Which of the following are true and which are false? Explain your answers.

**(a) Depth-first search always expands at least as many nodes as $A*$ search with an admissible heuristic.**

**Answer: False -** a lucky DFS might expand exactly d nodes to reach the goal. A* dominates any graph-search algorithm that is guaranteed to find an optimal solution.

**Explanation:** This is false for two reasons. First, while DFS could theoretically expand only d nodes with perfect luck, this is highly unlikely since DFS lacks guidance toward shorter paths and typically explores deeper unnecessarily. Second, A* doesn't dominate all optimal algorithmsonly those using the same heuristic information. Other optimal methods might use different knowledge sources that make them more efficient for specific problems. A*'s dominance applies only to best-first search with admissible heuristics, not all optimal search algorithms.

<br>

**(b) h(n) = 0 is an admissible heuristic for the 8-puzzle.**

**Answer: True-** h(n) = 0 is always an admissible heuristic, since costs are nonnegative.

**Explanation:** This statement is correct. A heuristic is admissible if it never overestimates the true cost to reach the goal. Since h(n) = 0 always returns zero and actual path costs in the 8-puzzle are always positive (each move has a cost  1), the heuristic never exceeds the true cost. While h(n) = 0 provides no useful guidance and makes A* equivalent to uniform-cost search, it remains technically admissible because it consistently underestimates the actual distance to the goal.
