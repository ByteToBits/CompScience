### **SIFT: Scale Invariant Feature Transform**

SIFT was introduced by David Lowe in *"Distinctive Image Features from Scale-Invariant Keypoints"* (International Journal of Computer Vision, 2004). It detects interest points that are invariant to scale and rotation, then describes each one with a distinctive feature vector that can be used for matching across images.

<br>

### **SIFT Descriptor**

#### **Step 1: Find scale-invariant interest points**

We first detect interest points at a particular scale using the DoG method described earlier. The corresponding **Gaussian-blurred image** at that scale is then selected as the working image for building the descriptor.

#### **Step 2: Extract a local patch**

Around each keypoint's neighbourhood, compute the **gradient magnitude and orientation** on the Gaussian-smoothed image at the scale of the keypoint. A **16 × 16 pixel patch** is extracted centred on the keypoint.

#### **Step 3: Build orientation histograms**

A histogram of gradient orientations is computed over the 16 × 16 patch using a **bin width of 10 degrees** (36 bins covering 360°).

#### **Step 4: Assign dominant orientation**

The orientation giving the **peak in the histogram** is chosen as the keypoint's dominant orientation. All subsequent computations are performed relative to this direction, making the descriptor **rotation invariant**.

#### **Step 5: Construct the descriptor vector**

The 16 × 16 patch is divided into a grid of **4 × 4 blocks**. For each block, a histogram with **8 orientation bins** is computed. This gives:

$$4 \times 4 \times 8 = 128 \text{ values}$$

These 128 values are concatenated into a single **128-dimensional feature vector** — the SIFT descriptor.

#### **Step 6: Normalisation for illumination invariance**

To handle changes in brightness and contrast, the 128-D vector is **normalised to unit length**:

$$f_i = \frac{f_i}{\|f_i\|}$$

For further photometric invariance, values are **clipped to a maximum of 0.2**, and the vector is **normalised again**. This prevents any single gradient direction from dominating the descriptor due to extreme lighting conditions.

### **Self-Similarity Descriptor**

Sometimes images of the same object share no common low-level properties like colour, texture, or edges — but they **do** share a similar geometric layout of **local internal self-similarities**. For example, a heart shape drawn in sand, formed in latte art, and painted on a wall all look completely different in terms of pixel values, yet share the same structural pattern.

The **Self-Similarity Descriptor** (Shechtman and Irani, ICCV 2007) captures this by measuring how a local patch relates to its surrounding region rather than describing the patch content directly.

#### **How it works**

1. Take an **image patch** (a small region around a point of interest)
2. Correlate that patch with the surrounding **image region** to produce a **correlation surface** — this shows where similar patterns appear nearby
3. Convert the correlation surface into a compact **image descriptor** using a log-polar binning scheme (a circular grid that is denser near the centre and coarser further out)

The key insight is that even when two images of the same object look completely different at the pixel level, the **pattern of local self-similarities** — how each small patch relates to its neighbours — remains consistent. This makes the descriptor robust to drastic changes in appearance across different depictions of the same object.

> *Reference: Shechtman and Irani, "Matching Local Self-Similarities across Images and Videos", ICCV 2007.*
