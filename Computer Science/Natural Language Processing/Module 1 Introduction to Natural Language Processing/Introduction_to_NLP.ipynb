### **Syntactic Tasks**

#### **2.1) Word Segmentation and Morphological Analysis**

**Word Segmentation:** Word Segmentation is the process to breaking up a **string of characters** into a **sequence of words**. In English, words can be seperated via the white-spaces or with some special characters like *, ; . - : ( )*. For example, *headsup.com* can be interpreted as 'heads up .com' or 'head sup .com'. 

**Morphological Analysis**: It is the process of **Segmenting Words** into morphemes. Where morpheme is the smallest linguistics unit that has semantic meaning (Example, 'carry', 'pre','ly') and morphology is the study of linguistics that concerns itself with the internal structure of word. Some examples are listed as follows: 

1) Carried = Carry + ed (past tense)
2) Independently = In + (depend + ent) + ly
3) Unlockable = Un + (lock + able)

<br>

#### **2.2) Part-of-speech (POS) Tagging and Phrase Chunking**

**Part-of-Speech (POS) Tagging:** POS tagging is the process of labeling each word in a text with its corresponding grammatical category (noun, verb, adjective, etc). This technique is useful for subsequent syntactic parsing and word sense disambiguation. The most popular POS Tagging table is the Penn Treebank 36 Part-of-Speech (POS) Tags.

| Penn      | PRP         | VBD    | DT           | NN        | IN            | NNS       |

### **Semantic Tasks**

#### **3.1) Word Sense Disambiguation**

Words in natural language can have several different meanings. Hence, the correct sense of each ambigous word in a sentence must be determined and that is where the need for **word sense disambiguation** comes into play. For example, the word **plant** can be used in two different ways: 

1) The industrial **plant** produces process utilities for the pharmaceutical company.
2) They intend to **plant** the seeds of democracy within the populace. 

<br>

#### **3.2) Semantic Role Labelling (SRL)**

This technique is also called case role analysis, thematic analysis, and shallow semantic parsing. For each clause, determine the semantic role played by each noun phrase that is an argument to the verb.

| Role        | Meaning                                  | Example                  |
|-------------|------------------------------------------|--------------------------|
| Agent       | The doer/initiator of the action         | John (drove)             |
| Patient     | The entity affected by the action        | Mary, the window         |
| Source      | Where the action originates from         | Austin                   |
| Destination | Where the action is directed toward      | Dallas                   |
| Instrument  | The tool/means used to perform the action| Toyota Prius, the hammer |

#### Sentence 1: "John drove Mary from Austin to Dallas in his Toyota Prius"

| Word/Phrase        | John     | drove | Mary    | from      | Austin  | to          | Dallas      | in            | his Toyota Prius |
|--------------------|----------|-------|---------|-----------|---------|-------------|-------------|---------------|------------------|
| Semantic Role      | Agent    |      | Patient |          | Source  |            | Destination |              | Instrument       |


##### Sentence 2: "The hammer broke the window"

#### **3.3) Semantic Parsing**

A semantic parser maps a natural-language sentence to a complete, detailed semantic representation (logical form). The logical form is a machine-understandable representation of the sentence meaning. Semantic parsing can thus be understood as extracting the precise meaning of a sentence. For many applications, the desired output is immediately executable by another program. The following example shows the mapping of an English database query to Prolog:

**Query:** How many cities are there in the US?

**Parse:** answer(A, count(B, (city(B), loc(B, C), const(C, countryid(USA))), A))

<br>

#### **3.4) Textual Entailment**

Textual Entailment is the task of determining if one sentence (T) logically implies another sentence (H). If a person reading T would reasonably conclude that H is true, then T entails H.

#### Example 1:

| Text (T)                                                                                                              | Hypothesis (H)        | Entailment |
|-----------------------------------------------------------------------------------------------------------------------|-----------------------|------------|
| Eyeing the huge market potential, currently led by Google, Yahoo took over search company Overture Services Inc last year. | Yahoo bought Overture | TRUE       |


#### Example 2:

| Text (T)                                                                                                              | Hypothesis (H)        | Entailment |
|-----------------------------------------------------------------------------------------------------------------------|-----------------------|------------|
| Microsoft's rival Sun Microsystems Inc. bought Star Office last month and plans to boost its development as a Web-based device running over the Net on personal computers and Internet appliances. | Sun bought Star Office   | TRUE       |

### **Pragmatics/Discourse Tasks**

#### **4.1 Anaphora/Co-reference Resolution**

**Purpose:** Identifies which different words or phrases in a text refer to the same underlying entity (person, object, concept).

#### Example 1: Simple Co-reference

**Sentence:** "John put the carrot on the plate and ate it."

**Reasoning:** "It" must refer to "carrot" because carrots are edible and plates are not. "Ate it" implies food.

| Phrase     | Refers to Entity |
|------------|------------------|
| John       | Entity 1 (John)  |
| the carrot | Entity 2 (carrot)|
| the plate  | Entity 3 (plate) |
| it         | Entity 2 (carrot)|

<br>

#### **4.2. Ellipsis Resolution**

**Purpose:** Fills in words or phrases that were omitted from sentences because they can be inferred from context.

#### Example

| Original (with ellipsis)                                                                      | Resolved (filled in)                                                                                    |
|-----------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------|
| Wise men talk because they have something to say; fools, because they have to say something. | Wise men talk because they have something to say; fools **talk** because they have to say something.   |

**Missing word:** "talk" (verb)

**Reasoning:** The verb "talk" is omitted after "fools" but is understood from the parallel structure in the first clause.

<br>

### **Approaches to Building NLP Systems**

#### **The Core Challenge**

To correctly understand language, systems need multiple types of knowledge:

| Knowledge Type   | What it provides              | Example                                                      |
|------------------|-------------------------------|--------------------------------------------------------------|
| **Syntax**       | Grammatical structure rules   | An agent is typically the subject of the verb                |
| **Semantics**    | Word meanings and categories  | Michael/Ellen = people, Austin = city, Toyota = company      |
| **Pragmatics**   | Context and implied meaning   | Understanding references and discourse                       |
| **World Knowledge** | Real-world facts           | Credit cards require users to pay financial interest         |

<br>

#### **Approach 1: Manual Knowledge Acquisition (Traditional)**

**Method:** Human experts manually write rules and formalize all linguistic knowledge the system needs.

**Problems:**
- Extremely time-consuming and expensive
- Error-prone and difficult
- Language has countless exceptions and irregularities
- Systems were **brittle** (not robust, broke easily on unexpected input)
- Limited capabilities

<br>

#### **Approach 2: Automatic Learning (Modern)**

**Method:** Use machine learning to automatically learn patterns from large collections of annotated text (corpora).



##### **How It Works:**

1. Humans annotate example documents (label parts of speech, entities, etc.)
2. Machine learning algorithms study these examples
3. System builds statistical/probabilistic knowledge
4. System automatically processes new, unseen text

##### **Advantages:**

- Large amounts of electronic text are now available
- Annotating corpora is easier and requires less expertise than manual knowledge engineering
- Learning algorithms can handle large amounts of data and produce accurate probabilistic knowledge
- Probabilistic knowledge allows **robust processing** that handles both linguistic regularities and exceptions

<br>

#### **The Importance of Probability**

Language is full of **ambiguity**. Without probability, systems get confused by unlikely but technically valid interpretations.

#### Example 1: Spurious Ambiguity

**Sentence:** "The a are of I"

**Why it's technically valid English:**

| Word | Part of Speech | Meaning                        |
|------|----------------|--------------------------------|
| The  | Determiner     | Determiner                     |
| a    | Adjective      | The letter A                   |
| are  | Noun           | Unit of area (like hectare)    |
| of   | Preposition    | Preposition                    |
| I    | Noun           | The letter I                   |

#### Example 2: Multiple Interpretations

**Sentence:** "Time flies like an arrow"

**Possible interpretations:**

1. Time passes quickly *(normal meaning)*
2. Insects called "time flies" are fond of a particular arrow
3. A command: measure insects' speed in the manner that an arrow would
4. Other interpretations...

**Solution:** Statistical methods calculate which interpretation is **most likely** based on:
- How often words appear together
- Common grammatical patterns
- Contextual evidence

This is why "Time passes quickly" wins over "insects fond of arrows."
