### **Syntactic Tasks**

#### **2.1) Word Segmentation and Morphological Analysis**

**Word Segmentation:** Word Segmentation is the process to breaking up a **string of characters** into a **sequence of words**. In English, words can be seperated via the white-spaces or with some special characters like *, ; . - : ( )*. For example, *headsup.com* can be interpreted as 'heads up .com' or 'head sup .com'. 

**Morphological Analysis**: It is the process of **Segmenting Words** into morphemes. Where morpheme is the smallest linguistics unit that has semantic meaning (Example, 'carry', 'pre','ly') and morphology is the study of linguistics that concerns itself with the internal structure of word. Some examples are listed as follows: 

1) Carried = Carry + ed (past tense)
2) Independently = In + (depend + ent) + ly
3) Unlockable = Un + (lock + able)

<br>

#### **2.2) Part-of-speech (POS) Tagging and Phrase Chunking**

**Part-of-Speech (POS) Tagging:** POS tagging is the process of labeling each word in a text with its corresponding grammatical category (noun, verb, adjective, etc). This technique is useful for subsequent syntactic parsing and word sense disambiguation.

**Phrase Chunking (also called shallow parsing):** It is the process of grouping individual words with their POS tags into meaningful chunks or phrases  but without going as deep as full syntactic parsing.

| Chunk | Name                  | Example          |
|-------|-----------------------|------------------|
| NP    | Noun Phrase           | "the big dog"    |
| VP    | Verb Phrase           | "is running fast"|
| PP    | Prepositional Phrase  | "on the table"   |
| ADJP  | Adjective Phrase      | "very happy"     |
| ADVP  | Adverb Phrase         | "quite slowly"   |

Syntactic Parsing is the process of analyzing a sentence to produce a parse tree (also called a syntax tree) that shows the grammatical structure and hierarchical relationships between all the words and phrases in a sentence.

#### **3.1) Word Sense Disambiguation**

Words in natural language can have several different meanings. For example, the word **plant** can be used in two different ways: 

1) The Industrial **Plant** produces process utilities for the pharmaceutical company.
2) They intend to **plant** the seeds of democracy within the populace.
