### **Question 1: Eigenvalues and Eigenvectors**

**Q1) Find all eigenvalues and their corresponding eigenvectors of A.**

$$A = \begin{bmatrix} -17 & 7 & 15 \\ -67 & 53 & 21 \\ -220 & 20 & 108 \end{bmatrix}$$

<br>

**Answer:** 
- **Eigenvalues:** $\lambda_1 = 72$, $\lambda_2 = 48$, $\lambda_3 = 24$
- **Corresponding Eigenvectors:** $\begin{bmatrix} 1 \\ 2 \\ 5 \end{bmatrix}$, $\begin{bmatrix} 1 \\ 5 \\ 2 \end{bmatrix}$, $\begin{bmatrix} 2 \\ 1 \\ 5 \end{bmatrix}$ respectively.

<br>

*Explanation: To find eigenvalues and eigenvectors, we solve the characteristic equation and then find the corresponding eigenvectors.*

**Step 1:** Find the characteristic polynomial by solving $\det(A - \lambda I) = 0$

$A - \lambda I = \begin{bmatrix} -17-\lambda & 7 & 15 \\ -67 & 53-\lambda & 21 \\ -220 & 20 & 108-\lambda \end{bmatrix}$

**Step 2:** Calculate the characteristic polynomial:
$\det(A - \lambda I) = (-17-\lambda)[(53-\lambda)(108-\lambda) - 21 \times 20] - 7[(-67)(108-\lambda) - 21(-220)] + 15[(-67)(20) - (53-\lambda)(-220)]$

**Step 3:** Solve the characteristic equation to get eigenvalues:
$\lambda_1 = 72, \quad \lambda_2 = 48, \quad \lambda_3 = 24$

**Step 4:** Find eigenvector for $\lambda_1 = 72$:
Solve $(A - 72I)\mathbf{v} = \mathbf{0}$:
$\begin{bmatrix} -89 & 7 & 15 \\ -67 & -19 & 21 \\ -220 & 20 & 36 \end{bmatrix}\begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$
This gives eigenvector $\mathbf{v_1} = \begin{bmatrix} 1 \\ 2 \\ 5 \end{bmatrix}$

**Step 5:** Find eigenvector for $\lambda_2 = 48$:
Solve $(A - 48I)\mathbf{v} = \mathbf{0}$:
$\begin{bmatrix} -65 & 7 & 15 \\ -67 & 5 & 21 \\ -220 & 20 & 60 \end{bmatrix}\begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$
This gives eigenvector $\mathbf{v_2} = \begin{bmatrix} 1 \\ 5 \\ 2 \end{bmatrix}$

**Step 6:** Find eigenvector for $\lambda_3 = 24$:
Solve $(A - 24I)\mathbf{v} = \mathbf{0}$:
$\begin{bmatrix} -41 & 7 & 15 \\ -67 & 29 & 21 \\ -220 & 20 & 84 \end{bmatrix}\begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$
This gives eigenvector $\mathbf{v_3} = \begin{bmatrix} 2 \\ 1 \\ 5 \end{bmatrix}$

### **Question 2: Linear System Analysis**

**Q2a) Find the augmented matrix for the systems of equations.**

$$-17x + 7y + 15z = 1$$
$$-67x + 53y + 21z = 3$$
$$-220x + 20y + az = b$$

<br>

**Answer:** 
$$\left[\begin{array}{ccc|c} -17 & 7 & 15 & 1 \\ -67 & 53 & 21 & 3 \\ -220 & 20 & a & b \end{array}\right]$$

<br>

**Q2b) Given the system of linear equations above, find values a and b such that:**
- i. the system has a unique solution
- ii. the system has infinitely many solutions  
- iii. the system has no solutions

<br>

**Answer:**
- **i) Unique solution:** $a \neq \frac{2203200}{7344}$
- **ii) Infinitely many solutions:** $a = \frac{2203200}{7344}$, $b = \frac{114240}{7344}$
- **iii) No solutions:** $a = \frac{2203200}{7344}$, $b \neq \frac{114240}{7344}$

<br>

*Explanation: The solution depends on the rank of the coefficient matrix and augmented matrix.*

**Step 1:** Form the augmented matrix:
$\left[\begin{array}{ccc|c} -17 & 7 & 15 & 1 \\ -67 & 53 & 21 & 3 \\ -220 & 20 & a & b \end{array}\right]$

**Step 2:** Perform row operations to reduce the matrix:
After row reduction, we get a system where the third row becomes:
$0 \cdot x + 0 \cdot y + (a - \frac{2203200}{7344}) \cdot z = b - \frac{114240}{7344}$

**Step 3:** Analyze the conditions:

**For unique solution:** The coefficient matrix must have full rank (rank = 3)
- This occurs when $a \neq \frac{2203200}{7344}$
- The third equation provides an independent constraint

**For infinitely many solutions:** rank(A) = rank([A|b]) = 2 < 3
- This occurs when $a = \frac{2203200}{7344}$ AND $b = \frac{114240}{7344}$
- The third equation becomes $0 = 0$ (always true)

**For no solutions:** rank(A) < rank([A|b])
- This occurs when $a = \frac{2203200}{7344}$ AND $b \neq \frac{114240}{7344}$
- The third equation becomes $0 = \text{non-zero}$ (contradiction)

### **Question 3: Vector Span**

**Q3) Describe the values that a, b, and c must satisfy for the vector $\begin{bmatrix} a \\ b \\ c \end{bmatrix}$ to be in the span of the following set:**

$$\left\{\begin{bmatrix} 1 \\ 3 \\ 5 \end{bmatrix}, \begin{bmatrix} 2 \\ 6 \\ 10 \end{bmatrix}\right\}$$

<br>

**Answer:** $a$ any value, $b = 3a$, $c = 5a$

<br>

*Explanation: For a vector to be in the span, it must be expressible as a linear combination of the spanning vectors.*

**Step 1:** Set up the equation $\begin{bmatrix} a \\ b \\ c \end{bmatrix} = \alpha \begin{bmatrix} 1 \\ 3 \\ 5 \end{bmatrix} + \beta \begin{bmatrix} 2 \\ 6 \\ 10 \end{bmatrix}$

**Step 2:** Notice that the second vector is $2 \times$ the first vector, so they are linearly dependent

**Step 3:** The span is effectively one-dimensional: $\text{span} = \left\{t\begin{bmatrix} 1 \\ 3 \\ 5 \end{bmatrix} : t \in \mathbb{R}\right\}$

**Step 4:** Therefore: $a = t$, $b = 3t = 3a$, $c = 5t = 5a$ for any real number $t$

### **Question 4: Matrix Operations**

**Q4) Suppose:**

$$A = \begin{bmatrix} 1 & 3 \\ 4 & 2 \end{bmatrix}, B = \begin{bmatrix} -1 & 3 \\ 3 & -2 \\ 2 & 4 \end{bmatrix}, C = \begin{bmatrix} 2 & 0 & 3 \\ -3 & 5 & 2 \end{bmatrix}, D = \begin{bmatrix} 5 & 2 & 3 \\ 2 & 5 & 7 \\ 3 & 4 & 3 \end{bmatrix}, E = \begin{bmatrix} 3 & 8 \\ 5 & 7 \end{bmatrix}$$

**Where the computations are possible, find:**
- a) $3A - 2E$
- b) $BC$  
- c) $CB + 4D$

<br>

**Answer:**

**a)** $3A - 2E = \begin{bmatrix} -1 & -7 \\ 2 & -8 \end{bmatrix}$

**b)** $BC = \begin{bmatrix} -11 & 15 & 3 \\ 12 & -10 & 5 \\ -8 & 20 & 14 \end{bmatrix}$

**c)** Undefined since $CB$ and $4D$ have different orders.

<br>

*Explanation: Matrix operations require compatible dimensions.*

**Step 1:** For part (a), compute $3A - 2E$:
Both $A$ and $E$ are $2 \times 2$ matrices, so this operation is valid.

$3A = 3\begin{bmatrix} 1 & 3 \\ 4 & 2 \end{bmatrix} = \begin{bmatrix} 3 & 9 \\ 12 & 6 \end{bmatrix}$

$2E = 2\begin{bmatrix} 3 & 8 \\ 5 & 7 \end{bmatrix} = \begin{bmatrix} 6 & 16 \\ 10 & 14 \end{bmatrix}$

$3A - 2E = \begin{bmatrix} 3-6 & 9-16 \\ 12-10 & 6-14 \end{bmatrix} = \begin{bmatrix} -1 & -7 \\ 2 & -8 \end{bmatrix}$

**Step 2:** For part (b), compute $BC$:
$B$ is $3 \times 2$ and $C$ is $2 \times 3$, so $BC$ will be $3 \times 3$.

$BC = \begin{bmatrix} -1 & 3 \\ 3 & -2 \\ 2 & 4 \end{bmatrix} \begin{bmatrix} 2 & 0 & 3 \\ -3 & 5 & 2 \end{bmatrix}$

$= \begin{bmatrix} (-1)(2) + (3)(-3) & (-1)(0) + (3)(5) & (-1)(3) + (3)(2) \\ (3)(2) + (-2)(-3) & (3)(0) + (-2)(5) & (3)(3) + (-2)(2) \\ (2)(2) + (4)(-3) & (2)(0) + (4)(5) & (2)(3) + (4)(2) \end{bmatrix}$

$= \begin{bmatrix} -2 - 9 & 0 + 15 & -3 + 6 \\ 6 + 6 & 0 - 10 & 9 - 4 \\ 4 - 12 & 0 + 20 & 6 + 8 \end{bmatrix} = \begin{bmatrix} -11 & 15 & 3 \\ 12 & -10 & 5 \\ -8 & 20 & 14 \end{bmatrix}$

**Step 3:** For part (c), analyze $CB + 4D$:
- $C$ is $2 \times 3$ and $B$ is $3 \times 2$, so $CB$ would be $2 \times 2$.
- $D$ is $3 \times 3$, so $4D$ would be $3 \times 3$.
- Since $CB$ ($2 \times 2$) and $4D$ ($3 \times 3$) have different dimensions, addition is undefined.

### **Question 5: Vector Form of Plane Equation**

**Q5) Given that a plane has equation $2x + 3y + 5z = 10$, rewrite the equation of this plane in vector form.**

<br>

**Answer:** $[2, 3, 5] \cdot [x - 0, y - 0, z - 2] = 0$

<br>

*Explanation: The vector form uses a normal vector and a point on the plane. Multiple correct forms are possible.*

**Step 1:** Identify the normal vector from coefficients: 
The equation $2x + 3y + 5z = 10$ has normal vector $\mathbf{n} = \begin{bmatrix} 2 \\ 3 \\ 5 \end{bmatrix}$ or $\mathbf{n} = 2\mathbf{i} + 3\mathbf{j} + 5\mathbf{k}$

**Step 2:** Find a reference point on the plane:
We can choose any point that satisfies the equation. Let's set $x = 0, y = 0$:
$2(0) + 3(0) + 5z = 10 \Rightarrow z = 2$
So point $P = (0, 0, 2)$ lies on the plane.

**Step 3:** Set up the vector form using the dot product:
For any point $Q = (x, y, z)$ on the plane, the vector $\overrightarrow{PQ} = (x-0, y-0, z-2)$ must be perpendicular to the normal vector $\mathbf{n}$.

**Step 4:** Apply the perpendicularity condition:
Since $\overrightarrow{PQ}$ lies on the plane, it's perpendicular to the normal vector:
$\mathbf{n} \cdot \overrightarrow{PQ} = 0$
$[2, 3, 5] \cdot [x - 0, y - 0, z - 2] = 0$

**Step 5:** Alternative vector forms:
We could choose different reference points on the plane:
- Using point $(5, 0, 0)$: $[2, 3, 5] \cdot [x - 5, y - 0, z - 0] = 0$
- Using point $(0, \frac{10}{3}, 0)$: $[2, 3, 5] \cdot [x - 0, y - \frac{10}{3}, z - 0] = 0$

All these forms are equivalent and represent the same plane in vector notation.

### **Question 6: Volume of Parallelepiped**

**Q6) Given the vectors $\mathbf{u} = \begin{bmatrix} 1 \\ 2 \\ 5 \end{bmatrix}$, $\mathbf{v} = \begin{bmatrix} 1 \\ 5 \\ 2 \end{bmatrix}$, $\mathbf{w} = \begin{bmatrix} 2 \\ 1 \\ 5 \end{bmatrix}$, find the volume of the parallelepiped spanned by $\mathbf{u}$, $\mathbf{v}$ and $\mathbf{w}$.**

<br>

**Answer:** $24$

<br>

*Explanation: The volume is the absolute value of the scalar triple product $\mathbf{u} \cdot (\mathbf{v} \times \mathbf{w})$.*

**Step 1:** Compute the cross product $\mathbf{v} \times \mathbf{w}$:
$$\mathbf{v} \times \mathbf{w} = \begin{bmatrix} 1 \\ 5 \\ 2 \end{bmatrix} \times \begin{bmatrix} 2 \\ 1 \\ 5 \end{bmatrix} = \begin{bmatrix} 5 \cdot 5 - 2 \cdot 1 \\ 2 \cdot 2 - 1 \cdot 5 \\ 1 \cdot 1 - 5 \cdot 2 \end{bmatrix} = \begin{bmatrix} 23 \\ -1 \\ -9 \end{bmatrix}$$

**Step 2:** Compute the scalar triple product:
$$\mathbf{u} \cdot (\mathbf{v} \times \mathbf{w}) = \begin{bmatrix} 1 \\ 2 \\ 5 \end{bmatrix} \cdot \begin{bmatrix} 23 \\ -1 \\ -9 \end{bmatrix} = 1(23) + 2(-1) + 5(-9) = 23 - 2 - 45 = -24$$

**Step 3:** Take absolute value for volume:
$$\text{Volume} = |-24| = 24$$

### **Question 7: Matrix Algebra Properties**

**Q7) Let $A = \begin{bmatrix} 1 & 2 \\ 4 & 3 \end{bmatrix}$, $B = \begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix}$. Compute:**
- a) $(A + B)^2$
- b) $A^2 + 2AB + B^2$
- c) $A^2 + AB + BA + B^2$

<br>

**Answer:**

**a)** $(A + B)^2 = \begin{bmatrix} 132 & 128 \\ 192 & 196 \end{bmatrix}$

**b)** $A^2 + 2AB + B^2 = \begin{bmatrix} 124 & 120 \\ 200 & 204 \end{bmatrix}$

**c)** $A^2 + AB + BA + B^2 = \begin{bmatrix} 132 & 128 \\ 192 & 196 \end{bmatrix}$ (same as part a)

<br>

*Explanation: This demonstrates that $(A + B)^2 \neq A^2 + 2AB + B^2$ for matrices due to non-commutativity.*

**Step 1:** Compute $A + B$:
$A + B = \begin{bmatrix} 1 & 2 \\ 4 & 3 \end{bmatrix} + \begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix} = \begin{bmatrix} 6 & 8 \\ 12 & 10 \end{bmatrix}$

**Step 2:** For part (a), compute $(A + B)^2$:
$(A + B)^2 = \begin{bmatrix} 6 & 8 \\ 12 & 10 \end{bmatrix} \begin{bmatrix} 6 & 8 \\ 12 & 10 \end{bmatrix}$

$= \begin{bmatrix} 6(6) + 8(12) & 6(8) + 8(10) \\ 12(6) + 10(12) & 12(8) + 10(10) \end{bmatrix} = \begin{bmatrix} 36 + 96 & 48 + 80 \\ 72 + 120 & 96 + 100 \end{bmatrix} = \begin{bmatrix} 132 & 128 \\ 192 & 196 \end{bmatrix}$

**Step 3:** For part (b), compute individual components:

$A^2 = \begin{bmatrix} 1 & 2 \\ 4 & 3 \end{bmatrix}^2 = \begin{bmatrix} 1+8 & 2+6 \\ 4+12 & 8+9 \end{bmatrix} = \begin{bmatrix} 9 & 8 \\ 16 & 17 \end{bmatrix}$

$B^2 = \begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix}^2 = \begin{bmatrix} 25+48 & 30+42 \\ 40+56 & 48+49 \end{bmatrix} = \begin{bmatrix} 73 & 72 \\ 96 & 97 \end{bmatrix}$

$AB = \begin{bmatrix} 1 & 2 \\ 4 & 3 \end{bmatrix} \begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix} = \begin{bmatrix} 5+16 & 6+14 \\ 20+24 & 24+21 \end{bmatrix} = \begin{bmatrix} 21 & 20 \\ 44 & 45 \end{bmatrix}$

$A^2 + 2AB + B^2 = \begin{bmatrix} 9 & 8 \\ 16 & 17 \end{bmatrix} + 2\begin{bmatrix} 21 & 20 \\ 44 & 45 \end{bmatrix} + \begin{bmatrix} 73 & 72 \\ 96 & 97 \end{bmatrix}$

$= \begin{bmatrix} 9+42+73 & 8+40+72 \\ 16+88+96 & 17+90+97 \end{bmatrix} = \begin{bmatrix} 124 & 120 \\ 200 & 204 \end{bmatrix}$

**Step 4:** For part (c), compute $BA$:
$BA = \begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix} \begin{bmatrix} 1 & 2 \\ 4 & 3 \end{bmatrix} = \begin{bmatrix} 5+24 & 10+18 \\ 8+28 & 16+21 \end{bmatrix} = \begin{bmatrix} 29 & 28 \\ 36 & 37 \end{bmatrix}$

$A^2 + AB + BA + B^2 = \begin{bmatrix} 9 & 8 \\ 16 & 17 \end{bmatrix} + \begin{bmatrix} 21 & 20 \\ 44 & 45 \end{bmatrix} + \begin{bmatrix} 29 & 28 \\ 36 & 37 \end{bmatrix} + \begin{bmatrix} 73 & 72 \\ 96 & 97 \end{bmatrix}$

$= \begin{bmatrix} 9+21+29+73 & 8+20+28+72 \\ 16+44+36+96 & 17+45+37+97 \end{bmatrix} = \begin{bmatrix} 132 & 128 \\ 192 & 196 \end{bmatrix}$

**Step 5:** Observe that $(A + B)^2 = A^2 + AB + BA + B^2$ but $(A + B)^2 \neq A^2 + 2AB + B^2$ because $AB \neq BA$ for matrices.

### **Question 8: Investment Problem**

**Q8) How do you invest $10,000 to earn exactly $600 a year interest, if investment option A pays 5% and investment option B pays 10% (higher risk)?**

<br>

**Answer:** Invest $8,000 in option A and $2,000 in option B.

<br>

*Explanation: This is a system of linear equations problem with constraints. We need to perform feasibility analysis first.*

**Step 1:** Feasibility Analysis:
Before solving, let's check if our target is achievable:
- **Option A alone:** $5\% \times \$10,000 = \$500$ (less than target $600)
- **Option B alone:** $10\% \times \$10,000 = \$1,000$ (more than target $600)

Since our target of $600 falls between the lower bound ($500) and upper bound ($1,000), the problem is feasible.

**Step 2:** Define variables:
- Let $x =$ amount invested in option A (5% return)
- Let $y =$ amount invested in option B (10% return)

**Step 3:** Set up the system of equations:
$x + y = 10,000 \quad \text{(total investment constraint)}$
$0.05x + 0.1y = 600 \quad \text{(target interest constraint)}$

**Step 4:** Convert to matrix form:
$\begin{bmatrix} 1 & 1 \\ 0.05 & 0.1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 10,000 \\ 600 \end{bmatrix}$

Or in augmented matrix form:
$\left[\begin{array}{cc|c} 1 & 1 & 10,000 \\ 0.05 & 0.1 & 600 \end{array}\right]$

**Step 5:** Solve the system using substitution:
From the first equation: $x = 10,000 - y$

Substitute into the second equation:
$0.05(10,000 - y) + 0.1y = 600$
$500 - 0.05y + 0.1y = 600$
$500 + 0.05y = 600$
$0.05y = 100$
$y = 2,000$

Therefore: $x = 10,000 - 2,000 = 8,000$

**Step 6:** Verification:
Check both constraints:
- **Total investment:** $8,000 + 2,000 = 10,000$ ✓
- **Target interest:** $0.05(8,000) + 0.1(2,000) = 400 + 200 = 600$ ✓

**Step 7:** Interpretation:
The solution tells us to split the $10,000 investment as follows:
- **$8,000 in option A** (lower risk, 5% return) → earns $400
- **$2,000 in option B** (higher risk, 10% return) → earns $200
- **Total annual interest:** $400 + $200 = $600 exactly

### **Question 9: Cramer's Rule**

**Q9) Use Cramer's rule to solve the following equations, showing all steps:**

$$-17x + 7y + 15z = 1$$
$$-67x + 53y + 21z = 3$$
$$-220x + 20y + 108z = 5$$

<br>

**Answer:** $x = \frac{696}{82944}$, $y = \frac{3768}{82944}$, $z = \frac{4560}{82944}$

<br>

*Explanation: Cramer's rule uses determinants to solve systems: $x = \frac{\det(A_1)}{\det(A)}$, $y = \frac{\det(A_2)}{\det(A)}$, $z = \frac{\det(A_3)}{\det(A)}$*

**Step 1:** Calculate the main determinant $\det(A)$:
$$\det(A) = \begin{vmatrix} -17 & 7 & 15 \\ -67 & 53 & 21 \\ -220 & 20 & 108 \end{vmatrix} = 82944$$

**Step 2:** Calculate $\det(A_1)$ (replace first column with constants):
$$\det(A_1) = \begin{vmatrix} 1 & 7 & 15 \\ 3 & 53 & 21 \\ 5 & 20 & 108 \end{vmatrix} = 696$$

**Step 3:** Calculate $\det(A_2)$ (replace second column with constants):
$$\det(A_2) = \begin{vmatrix} -17 & 1 & 15 \\ -67 & 3 & 21 \\ -220 & 5 & 108 \end{vmatrix} = 3768$$

**Step 4:** Calculate $\det(A_3)$ (replace third column with constants):
$$\det(A_3) = \begin{vmatrix} -17 & 7 & 1 \\ -67 & 53 & 3 \\ -220 & 20 & 5 \end{vmatrix} = 4560$$

**Step 5:** Apply Cramer's rule:
$$x = \frac{\det(A_1)}{\det(A)} = \frac{696}{82944}$$
$$y = \frac{\det(A_2)}{\det(A)} = \frac{3768}{82944}$$
$$z = \frac{\det(A_3)}{\det(A)} = \frac{4560}{82944}$$
