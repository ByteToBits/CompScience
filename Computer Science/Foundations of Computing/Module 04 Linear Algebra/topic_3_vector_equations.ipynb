### **Equation of a Straight Line**

#### **Cartesian Forms in 2D Space**

The equation of a straight line can be expressed in several Cartesian forms:

#### Standard Form
$$y = mx + c$$
where $m$ is the gradient (slope) of the straight line.

#### General Form
$$ax + by = c$$

Both are Cartesian forms of the equation of a straight line in 2-dimensional space.

#### **Vector Form in 2D Space**

We can also express the equation of a straight line using a vector equation:
$$\mathbf{r}(t) = \mathbf{a} + t\mathbf{v}$$

where:
- $\mathbf{a}$ and $\mathbf{v}$ are vectors in 2-dimensional space
- $t$ is a free variable representing any real number (parameter)

#### **Example: Converting from Cartesian to Vector Form**

Given the Cartesian equation: $y = 2x + 5$

**Step 1: Find a point on the line**
When $x = 0$, $y = 5$
So $(0, 5)$ is a point on the line.
Position vector: $\mathbf{a} = \begin{bmatrix} 0 \\ 5 \end{bmatrix}$

**Step 2: Find the direction vector**
The gradient is $m = 2$, meaning for each unit increase in $x$, $y$ increases by 2 units.
Direction vector: $\mathbf{v} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$

**Step 3: Write the vector equation**
$$\mathbf{r}(t) = \begin{bmatrix} 0 \\ 5 \end{bmatrix} + t\begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad t \in \mathbb{R}$$

#### **Alternative Representations**

**Using a different point:**
We could choose the point where $y = 0$ and $x = -2.5$:
$$\mathbf{r}(t) = \begin{bmatrix} -2.5 \\ 0 \end{bmatrix} + t\begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad t \in \mathbb{R}$$

**Using a scaled direction vector:**
Let $\mathbf{v} = \begin{bmatrix} 10 \\ 20 \end{bmatrix}$ (same direction, different magnitude):
$$\mathbf{r}(t) = \begin{bmatrix} -2.5 \\ 0 \end{bmatrix} + t\begin{bmatrix} 10 \\ 20 \end{bmatrix}, \quad t \in \mathbb{R}$$

All these represent the same line!

#### 3. Equation of a Line in 3D Space

#### Cartesian Form (Symmetric Equations)
$$\frac{x - x_0}{a} = \frac{y - y_0}{b} = \frac{z - z_0}{c}$$

where:
- $(x_0, y_0, z_0)$ is a known point on the line
- $\begin{bmatrix} a \\ b \\ c \end{bmatrix}$ is the direction vector of the line

### **Geometric Interpretation**

All points on the line travel in the same direction as the direction of the line. 

If we choose a reference point $(x_0, y_0, z_0)$, then for all other points $(x, y, z)$ on the line, the ratios must satisfy:
$$x - x_0 : y - y_0 : z - z_0 = a : b : c$$

#### **Vector Form in 3D**
$$\mathbf{r}(t) = \begin{bmatrix} x_0 \\ y_0 \\ z_0 \end{bmatrix} + t\begin{bmatrix} a \\ b \\ c \end{bmatrix}, \quad t \in \mathbb{R}$$

### **Equation of a Plane**

#### **Cartesian Form**

The general equation of a plane in Cartesian form is:
$ax + by + cz = d$

where:
- The plane is a flat surface in 3D space
- $\mathbf{n} = \begin{bmatrix} a \\ b \\ c \end{bmatrix}$ is the **normal vector** to the plane
- A normal vector is perpendicular to the plane from all directions

#### **Degrees of Freedom Analysis**

The equation $ax + by + cz = d$ has:
- **3 unknowns** $(x, y, z)$
- **1 equation**
- Therefore: $3 - 1 = 2$ **degrees of freedom**

This means a plane is a **2-dimensional object** existing in 3-dimensional space.

Based on vector subspace properties, we need **two linearly independent vectors** to span this plane.

<br>

### **Vector Equations for a Plane**

There are two methods to construct a vector equation for a plane:

#### **Parametric Form (Using Two Direction Vectors)**

**Steps:**
1. Choose a reference point on the plane: $\mathbf{P_0} = (x_0, y_0, z_0)$
2. Identify two linearly independent vectors on the plane: $\mathbf{u}$ and $\mathbf{v}$
3. Construct the equation:

$\mathbf{r}(s,t) = \mathbf{P_0} + s\mathbf{u} + t\mathbf{v}$

where:
- $\mathbf{u}$ and $\mathbf{v}$ are linearly independent vectors lying in the plane
- $s, t \in \mathbb{R}$ are free parameters
- $\mathbb{R}$ is the set of all real numbers

#### **Normal Form (Using Normal Vector)**

**Steps:**
1. Construct the normal vector from the plane equation $ax + by + cz = d$:
   $\mathbf{n} = \begin{bmatrix} a \\ b \\ c \end{bmatrix}$
   
2. Choose a reference point on the plane: $\mathbf{P_0} = (x_0, y_0, z_0)$

3. Let $\mathbf{r} = (x, y, z)$ be any point on the plane

4. Construct the equation:
   $(\mathbf{r} - \mathbf{P_0}) \cdot \mathbf{n} = 0$

**Geometric Interpretation:**
- $(\mathbf{r} - \mathbf{P_0})$ produces a vector lying on the plane
- Since $\mathbf{n}$ is perpendicular to the plane from all directions
- $\mathbf{n}$ is perpendicular to $(\mathbf{r} - \mathbf{P_0})$
- Therefore, their dot product equals zero

**Method 2** is more commonly used than Method 1. When a question asks to "find the equation of the plane in vector form," it typically means the normal form specified in Method 2.

### Summary Comparison

| Object | Dimension | Spanning Vectors | Equation Form |
|--------|-----------|------------------|---------------|
| Line | 1D | 1 direction vector | $\mathbf{r}(t) = \mathbf{a} + t\mathbf{v}$ |
| Plane | 2D | 2 independent vectors | $\mathbf{r}(s,t) = \mathbf{P_0} + s\mathbf{u} + t\mathbf{v}$ |
| Plane (Normal) | 2D | 1 normal vector | $(\mathbf{r} - \mathbf{P_0}) \cdot \mathbf{n} = 0$ |

### **Worksheet Examples**

#### **Vector Equation of a Straight Line**

**Q1) A straight line $l$ passes through 2 points $A(2, 5, 3)$ and $B(6, 0, 5)$. Find the vector equation of the straight line.**

*Explanation: To find the vector equation of a straight line passing through two points, we need a point on the line and a direction vector.*

$$\text{Given points:}$$
$$A = (2, 5, 3)$$
$$B = (6, 0, 5)$$

**Step 1: Find the direction vector $\overrightarrow{AB}$**

$$\overrightarrow{AB} = \vec{B} - \vec{A}$$

$$\overrightarrow{AB} = (6, 0, 5) - (2, 5, 3)$$

$$\overrightarrow{AB} = (6 - 2, 0 - 5, 5 - 3)$$

$$\overrightarrow{AB} = (4, -5, 2)$$

**Step 2: Write the vector equation of the line**

The general form of a vector equation for a straight line is:
$$\vec{r} = \vec{a} + \lambda\vec{d}$$

Where:
- $\vec{a}$ is the position vector of a known point on the line
- $\vec{d}$ is the direction vector
- $\lambda$ is a parameter

Using point $A(2, 5, 3)$ and direction vector $\overrightarrow{AB} = (4, -5, 2)$:

$$\vec{r} = (2, 5, 3) + \lambda(4, -5, 2)$$

**Step 3: Expand in component form**

$$\vec{r} = (2 + 4\lambda, 5 - 5\lambda, 3 + 2\lambda)$$

**Step 4: Express in unit vector notation**

$$\vec{r} = (2 + 4\lambda)\mathbf{i} + (5 - 5\lambda)\mathbf{j} + (3 + 2\lambda)\mathbf{k}$$

Therefore, the vector equation of the straight line is $\vec{r} = (2, 5, 3) + \lambda(4, -5, 2)$.

<br>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define points
A = np.array([2, 5, 3])
B = np.array([6, 0, 5])

# Direction vector
AB = B - A
print(f"Direction vector AB = {AB}")
print(f"Vector equation: r = {A} + Î»{AB}")

# Create line points using parameter Î»
lambda_vals = np.linspace(-2, 3, 100)  # Range of Î» values
line_points = np.array([A + lam * AB for lam in lambda_vals])

# Plot
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the line (shorter range)
lambda_short = np.linspace(-1, 2, 50)
line_short = np.array([A + lam * AB for lam in lambda_short])
ax.plot(line_short[:, 0], line_short[:, 1], line_short[:, 2], 
        'b-', linewidth=4, label='Line l')

# Plot points A and B (bigger)
ax.scatter(*A, color='red', s=200, label='A', edgecolor='black', linewidth=2)
ax.scatter(*B, color='green', s=200, label='B', edgecolor='black', linewidth=2)

# Add coordinate labels (bigger font, better positioning)
ax.text(A[0]+0.5, A[1]+0.5, A[2]+0.5, f'A({A[0]}, {A[1]}, {A[2]})', 
        fontsize=14, fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
ax.text(B[0]+0.5, B[1]+0.5, B[2]+0.5, f'B({B[0]}, {B[1]}, {B[2]})', 
        fontsize=14, fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

# Set better axis limits
ax.set_xlim(0, 8)
ax.set_ylim(-2, 7)
ax.set_zlim(1, 7)

# Larger labels
ax.set_xlabel('X', fontsize=14, fontweight='bold')
ax.set_ylabel('Y', fontsize=14, fontweight='bold')
ax.set_zlabel('Z', fontsize=14, fontweight='bold')
ax.set_title('Vector Equation: r = (2,5,3) + Î»(4,-5,2)', fontsize=16, fontweight='bold', pad=20)

# Bigger legend
ax.legend(fontsize=12, loc='upper right')

# Less cluttered grid
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
