### **Vectors**

A **vector** is a fundamental mathematical object characterized by two essential properties:
- **Direction**: The orientation in space
- **Magnitude**: The length or size of the vector

A vector in an $n$-dimensional space is represented as an $n$-tuple of real numbers.

### **Vector Notation and Representation**

Vectors can be represented in several ways:

**Row notation:** $[a, b]$ or $(a, b)$ for 2D vectors

**Column notation:** 
$$\begin{pmatrix} a \\ b \\ c \end{pmatrix}$$

**Examples:**
- 2D vector: $[a, b]$ moves $a$ units along the x-axis and $b$ units along the y-axis
- 3D vector: $[a, b, c]$ extends into three-dimensional space
- Specific example: $(1, 2, 3)$ represents a vector with components 1, 2, and 3

**Note:** There is potential ambiguity when using normal brackets $($ and $)$ since they are also used for coordinates. Square brackets $[$ and $]$ are preferred for vector notation.

### **Vector Magnitude (Length)**

The **magnitude** or **length** of a vector is computed using the Euclidean distance formula.

For a vector $\mathbf{v} = [a, b, c]$ in 3D space:

$$|\mathbf{v}| = \sqrt{a^2 + b^2 + c^2}$$

More generally, for an $n$-dimensional vector $\mathbf{v} = [v_1, v_2, \ldots, v_n]$:

$$|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$$

This formula assumes that each component of the vector lies in an axis system where all axes are orthogonal to each other (such as x-axis, y-axis, z-axis). We refer to these as **Euclidean vectors**.

### **Vector Equivalence**

**Important Property:** Vectors with the same magnitude and direction are considered identical, regardless of their starting position in space.

For example, three vectors in 2D space are the same vector if they have:
- The same length (magnitude)
- The same direction
- Different starting points do not affect vector equality

### **Linearly Independent Axes**

In any axis system, each axis must be **linearly independent** of the others. This means that no axis can be expressed as a linear combination of the other axes.

### **Three Equivalent Definitions of Linear Independence**

Given axes $x_1, x_2, \ldots, x_n$, the following statements are equivalent:

1. **No axis as linear combination:** Every axis cannot be expressed as a linear combination of all other axes.

2. **Uniqueness of representation:** For any constants $a_1, a_2, \ldots, a_n$ and $b_1, b_2, \ldots, b_n$:
   $$a_1x_1 + a_2x_2 + \cdots + a_nx_n = b_1x_1 + b_2x_2 + \cdots + b_nx_n$$
   if and only if $a_1 = b_1, a_2 = b_2, \ldots, a_n = b_n$

3. **Trivial null combination:** For any constants $a_1, a_2, \ldots, a_n$:
   $$a_1x_1 + a_2x_2 + \cdots + a_nx_n = 0$$
   if and only if $a_1 = 0, a_2 = 0, \ldots, a_n = 0$

### **Constants and Scalars**

A **constant** or **scalar** refers to any real number. Scalars are used to scale vectors (change their magnitude) without affecting their direction.

### **Dimensional Classification**

A vector is fundamentally a **1-dimensional object** regardless of how many components are used to describe it:

- **0-dimensional object:** A single point (no extent)
- **1-dimensional object:** Has length only (like a vector)
- **2-dimensional object:** Has area (like a plane or surface)
- **3-dimensional object:** Has volume (like a solid)

The number of components in a vector (2D, 3D, nD) describes the **space** in which the vector exists, not the **dimensionality** of the vector object itself. The vector remains a 1-dimensional entity with length and direction.

### **Vector Spaces**

### Definition
A **vector space** is a collection of vectors that can be added and multiplied by scalars. Every vector space has a dimension:
- 1D: straight line
- 2D: plane  
- 3D: three-dimensional space

Vectors in the same vector space are **closed** under:
- Scalar multiplication
- Vector addition

Vectors from different dimensions cannot be added or multiplied.

### **Scalar Multiplication**
If $\mathbf{a} = [a_1, a_2, \ldots, a_n]$ and $C$ is a real number:
$$C\mathbf{a} = [Ca_1, Ca_2, \ldots, Ca_n]$$

Example:

Original vector: $$\mathbf{v} = [2, 2]$$
$$2\mathbf{v} = 2[2, 2] = [4, 4]$$
$$3\mathbf{v} = 3[2, 2] = [6, 6]$$

### **Vector Addition**
If $\mathbf{a} = [a_1, a_2, \ldots, a_n]$ and $\mathbf{b} = [b_1, b_2, \ldots, b_n]$:
$$\mathbf{a} + \mathbf{b} = [a_1 + b_1, a_2 + b_2, \ldots, a_n + b_n]$$

**Important:** Only vectors of the same dimension can be added.
- $[1, 2] + [3, 4] = [4, 6]$ ✓
- $[1, 2] + [1, 2, 3]$ ✗ (illegal operation)

**Example:**

Vector addition: $$[3, 4] + [3, 0] = [3+3, 4+0] = [6, 4]$$
Magnitude calculations:

$$|\mathbf{u}| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$$
$$|\mathbf{v}| = \sqrt{3^2 + 0^2} = \sqrt{9} = 3$$
$$|\mathbf{u} + \mathbf{v}| = \sqrt{6^2 + 4^2} = \sqrt{36 + 16} = \sqrt{52}$$

### **Linear Combination**
A linear combination takes multiple vectors and combines them using scalar multiplication and addition.
For vectors $\mathbf{v_1}, \mathbf{v_2}, \ldots, \mathbf{v_n}$ and scalars $a_1, a_2, \ldots, a_n$:
$a_1\mathbf{v_1} + a_2\mathbf{v_2} + \cdots + a_n\mathbf{v_n}$

Steps: Multiply each vector by its scalar Add all the results together

Example: For $\mathbf{u} = [1, 2]$, $\mathbf{v} = [3, -1]$, and scalars $a = 2$, $b = -1$:
$2\mathbf{u} + (-1)\mathbf{v} = 2[1, 2] + (-1)[3, -1] = [2, 4] + [-3, 1] = [-1, 5]$

### **Linear Operations**
Any combination of scalar multiplication and vector addition on vectors from the same dimensional space. For Example, $[1, 2] + [1, 2, 3]$ is considered an illegal operation because they are not in the same dimensional space.

### **Vector Subspace**

**Definition:** A vector space $U$ is a **vector subspace** of vector space $V$ if:
- $U \subseteq V$ (every vector in $U$ is also in $V$)
- $U$ satisfies all vector space axioms under the same operations as $V$

**Subspace Test:** For $U \subseteq V$ to be a subspace, it must satisfy:
1. **Non-empty:** $\mathbf{0} \in U$
2. **Closed under addition:** $\forall u_1, u_2 \in U: u_1 + u_2 \in U$
3. **Closed under scalar multiplication:** $\forall u \in U, \forall c \in \mathbb{R}: cu \in U$

### **Span**

**Definition:** Let $S = \{v_1, v_2, v_3, \ldots, v_n\}$ be a set of vectors from the same vector space $V$.

$\text{Span}(S) = \text{Span}(v_1, v_2, \ldots, v_n) = \left\{ \sum_{i=1}^{n} c_i v_i : c_i \in \mathbb{R} \right\}$

The **span** of $S$ is the set of all possible **linear combinations** of vectors in $S$.

#### Key Properties of Span:
- $\text{Span}(S)$ is always a vector subspace of $V$
- $\text{Span}(S)$ is the **smallest** vector subspace containing all vectors in $S$
- $S \subseteq \text{Span}(S)$

#### Examples:

| Set $S$ | Vectors | Span Description |
|---------|---------|------------------|
| $S_1 = \{\mathbf{a}, \mathbf{b}\}$ | $\mathbf{a} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \mathbf{b} = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$ | $\text{Span}(S_1) = \mathbb{R}^2$ (entire plane) |
| $S_2 = \{\mathbf{c}, \mathbf{d}\}$ | $\mathbf{c} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}, \mathbf{d} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$ | $\text{Span}(S_2) = \mathbb{R}^2$ (entire plane) |

### **Basis**

**Definition:** Given a vector space $V$, a set $B = \{b_1, b_2, \ldots, b_n\}$ is a **basis** for $V$ if:

1. **Linear Independence:** The vectors in $B$ are linearly independent
2. **Spanning Property:** $\text{Span}(B) = V$

#### Properties of a Basis:

| Property | Description | Mathematical Statement |
|----------|-------------|----------------------|
| **Minimality** | Cannot remove any vector without losing the spanning property | If $B' \subset B$, then $\text{Span}(B') \neq V$ |
| **Maximality** | Cannot add any vector without losing linear independence | If $B' \supset B$ and $B' \subset V$, then $B'$ is linearly dependent |
| **Uniqueness of Representation** | Every vector in $V$ has a unique representation as a linear combination of basis vectors | $\forall v \in V, \exists! (c_1, \ldots, c_n): v = \sum_{i=1}^{n} c_i b_i$ |

#### Key Results:
- **Existence:** Every vector space has a basis
- **Dimension:** All bases of a finite-dimensional vector space have the same number of elements
- **Standard Basis:** For $\mathbb{R}^n$, the standard basis is $\{e_1, e_2, \ldots, e_n\}$ where $e_i$ has 1 in the $i$-th position and 0 elsewhere

## Relationships Summary

$\text{Vector Space V}$
↓
$\text{Subspace U ⊆ V}$
↓
$\text{Span(S) = smallest subspace containing set S}$
    ↓

### **Properties and Terminologies of Vector Space**

#### **Vector Addition Properties**

| Property | Mathematical Statement | Description |
|----------|----------------------|-------------|
| **Closure of Addition** | $\forall u, v \in V: u + v \in V$ | The sum of any two vectors in $V$ is also in $V$ |
| **Associativity of Addition** | $\forall u, v, w \in V: (u + v) + w = u + (v + w)$ | Vector addition is associative |
| **Commutativity** | $\forall u, v \in V: u + v = v + u$ | Vector addition is commutative |
| **Existence of Identity** | $\exists \mathbf{0} \in V: \forall u \in V, u + \mathbf{0} = u$ | There exists a zero vector |
| **Existence of Inverse** | $\forall u \in V, \exists (-u) \in V: u + (-u) = \mathbf{0}$ | Every vector has an additive inverse |

#### **Scalar Multiplication Properties**

| Property | Mathematical Statement | Description |
|----------|----------------------|-------------|
| **Closure of Scalar Multiplication** | $\forall u \in V, \forall c \in \mathbb{R}: cu \in V$ | Scalar multiples of vectors remain in $V$ |
| **Distributive Property 1** | $\forall u, v \in V, \forall c \in \mathbb{R}: c(u + v) = cu + cv$ | Scalars distribute over vector addition |
| **Distributive Property 2** | $\forall u \in V, \forall c, d \in \mathbb{R}: (c + d)u = cu + du$ | Scalar addition distributes over vectors |
| **Associativity of Scalar Multiplication** | $\forall u \in V, \forall c, d \in \mathbb{R}: (cd)u = c(du)$ | Scalar multiplication is associative |
| **Identity of Scalar Multiplication** | $\forall u \in V: 1 \cdot u = u$ | Multiplication by 1 leaves vectors unchanged |

### **Graphical representation of a Vector**

### Vectors in 2-Dimensional Space

**Setup: The Cartesian Plane**
Consider the $xy$-plane with:
- **Origin:** $O = (0, 0)$
- **Point P:** $P = (a, b)$  
- **Point Q:** $Q = (c, d)$

**Position Vectors:** A **position vector** is a vector that starts at the origin and points to a specific location.

For point $P = (a, b)$, we create the position vector:
$$\overrightarrow{OP} = \begin{bmatrix} a \\ b \end{bmatrix}$$

This vector $\overrightarrow{OP}$ can be visualized as:
- An arrow starting at the origin $(0, 0)$
- Ending at point $P(a, b)$
- Having horizontal component $a$ and vertical component $b$

Similarly, for point $Q = (c, d)$:
$$\overrightarrow{OQ} = \begin{bmatrix} c \\ d \end{bmatrix}$$

<br>

### **Vector Between Two Points**

What if we want to find the vector $\overrightarrow{PQ}$ that goes directly from point $P$ to point $Q$?

#### **Geometric Interpretation**

Imagine taking a "detour" path:
1. Start at origin $O$
2. Go to point $P$ (following $\overrightarrow{OP}$)
3. Then go from $P$ to $Q$ (following $\overrightarrow{PQ}$)

The **net effect** is the same as going directly from $O$ to $Q$.

#### Vector Addition Relationship

This geometric reasoning gives us:
$$\overrightarrow{OP} + \overrightarrow{PQ} = \overrightarrow{OQ}$$

#### Solving for $\overrightarrow{PQ}$

Rearranging the vector equation:
$$\overrightarrow{PQ} = \overrightarrow{OQ} - \overrightarrow{OP}$$

Substituting the coordinates:
$$\overrightarrow{PQ} = \begin{bmatrix} c \\ d \end{bmatrix} - \begin{bmatrix} a \\ b \end{bmatrix} = \begin{bmatrix} c - a \\ d - b \end{bmatrix}$$



<br>

### **Key Formula: Vector Between Two Points**

For any two points $P(a, b)$ and $Q(c, d)$ in 2D space:

$$\boxed{\overrightarrow{PQ} = \begin{bmatrix} c - a \\ d - b \end{bmatrix}}$$

#### Interpretation
- **Horizontal displacement:** $c - a$ (change in $x$-coordinate)
- **Vertical displacement:** $d - b$ (change in $y$-coordinate)

<br>

### **Extension to 3-Dimensional Space**

In 3-dimensional space, we work with the $xyz$-coordinate system:
- **Origin:** $O = (0, 0, 0)$
- **Point P:** $P = (a, b, c)$
- **Point Q:** $Q = (d, e, f)$

#### Position Vectors in 3D

$$\overrightarrow{OP} = \begin{bmatrix} a \\ b \\ c \end{bmatrix}, \quad \overrightarrow{OQ} = \begin{bmatrix} d \\ e \\ f \end{bmatrix}$$

#### Vector Between Points in 3D

Using the same principle:
$$\overrightarrow{PQ} = \overrightarrow{OQ} - \overrightarrow{OP} = \begin{bmatrix} d \\ e \\ f \end{bmatrix} - \begin{bmatrix} a \\ b \\ c \end{bmatrix} = \begin{bmatrix} d - a \\ e - b \\ f - c \end{bmatrix}$$

### **Summary of Key Concepts**

| Concept | 2D Formula | 3D Formula |
|---------|------------|------------|
| **Position Vector** | $\overrightarrow{OP} = \begin{bmatrix} a \\ b \end{bmatrix}$ | $\overrightarrow{OP} = \begin{bmatrix} a \\ b \\ c \end{bmatrix}$ |
| **Vector Between Points** | $\overrightarrow{PQ} = \begin{bmatrix} c-a \\ d-b \end{bmatrix}$ | $\overrightarrow{PQ} = \begin{bmatrix} d-a \\ e-b \\ f-c \end{bmatrix}$ |
| **Vector Addition** | $\overrightarrow{OP} + \overrightarrow{PQ} = \overrightarrow{OQ}$ | $\overrightarrow{OP} + \overrightarrow{PQ} = \overrightarrow{OQ}$ |

### **Worksheet Examples**

**Q1) Relative to a fixed origin, the respective position vectors of three points A, B and C are (3, 2, 9), (-5, 11, 6) and (4, 0, -8). What are the vectors $\overrightarrow{AB}$ and $\overrightarrow{AC}$?**

*Explanation: To find vectors between points, we subtract the position vector of the starting point from the position vector of the ending point.*

$$\text{Given position vectors:}$$
$$\vec{A} = (3, 2, 9)$$
$$\vec{B} = (-5, 11, 6)$$
$$\vec{C} = (4, 0, -8)$$

**Step 1: Find vector $\overrightarrow{AB}$**

$$\overrightarrow{AB} = \vec{B} - \vec{A}$$

$$\overrightarrow{AB} = (-5, 11, 6) - (3, 2, 9)$$

$$\overrightarrow{AB} = (-5 - 3, 11 - 2, 6 - 9)$$

$$\overrightarrow{AB} = (-8, 9, -3) = -8\mathbf{i} + 9\mathbf{j} - 3\mathbf{k}$$

**Step 2: Find vector $\overrightarrow{AC}$**

$$\overrightarrow{AC} = \vec{C} - \vec{A}$$

$$\overrightarrow{AC} = (4, 0, -8) - (3, 2, 9)$$

$$\overrightarrow{AC} = (4 - 3, 0 - 2, -8 - 9)$$

$$\overrightarrow{AC} = (1, -2, -17) = \mathbf{i} - 2\mathbf{j} - 17\mathbf{k}$$

**Answer: Therefore, $\overrightarrow{AB} = (-8, 9, -3) = -8\mathbf{i} + 9\mathbf{j} - 3\mathbf{k}$ and $\overrightarrow{AC} = (1, -2, -17) = \mathbf{i} - 2\mathbf{j} - 17\mathbf{k}$.**

<br>


````python
import matplotlib.pyplot as plt
import numpy as np

# Define points
A = np.array([3, 2, 9])
B = np.array([-5, 11, 6])
C = np.array([4, 0, -8])

# Calculate vectors
AB = B - A
AC = C - A

print(f"AB = {AB} = {AB[0]}i + {AB[1]}j + {AB[2]}k")
print(f"AC = {AC} = {AC[0]}i + {AC[1]}j + {AC[2]}k")

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Points
ax.scatter(*A, color='red', s=100, label='A')
ax.scatter(*B, color='blue', s=100, label='B')
ax.scatter(*C, color='green', s=100, label='C')

# Add coordinate labels
ax.text(A[0]+0.5, A[1]+0.5, A[2]+0.5, f'A({A[0]}, {A[1]}, {A[2]})', fontsize=10)
ax.text(B[0]+0.5, B[1]+0.5, B[2]+0.5, f'B({B[0]}, {B[1]}, {B[2]})', fontsize=10)
ax.text(C[0]+0.5, C[1]+0.5, C[2]+0.5, f'C({C[0]}, {C[1]}, {C[2]})', fontsize=10)

# Vectors from A (using simple lines instead of arrows)
ax.plot([A[0], B[0]], [A[1], B[1]], [A[2], B[2]], color='purple', linewidth=3, label='AB')
ax.plot([A[0], C[0]], [A[1], C[1]], [A[2], C[2]], color='orange', linewidth=3, label='AC')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
````
