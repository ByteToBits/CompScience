### **Vectors**

A **vector** is a fundamental mathematical object characterized by two essential properties:
- **Direction**: The orientation in space
- **Magnitude**: The length or size of the vector

A vector in an $n$-dimensional space is represented as an $n$-tuple of real numbers.

### **Vector Notation and Representation**

Vectors can be represented in several ways:

**Row notation:** $[a, b]$ or $(a, b)$ for 2D vectors

**Column notation:** 
$$\begin{pmatrix} a \\ b \\ c \end{pmatrix}$$

**Examples:**
- 2D vector: $[a, b]$ moves $a$ units along the x-axis and $b$ units along the y-axis
- 3D vector: $[a, b, c]$ extends into three-dimensional space
- Specific example: $(1, 2, 3)$ represents a vector with components 1, 2, and 3

**Note:** There is potential ambiguity when using normal brackets $($ and $)$ since they are also used for coordinates. Square brackets $[$ and $]$ are preferred for vector notation.

### **Vector Magnitude (Length)**

The **magnitude** or **length** of a vector is computed using the Euclidean distance formula.

For a vector $\mathbf{v} = [a, b, c]$ in 3D space:

$$|\mathbf{v}| = \sqrt{a^2 + b^2 + c^2}$$

More generally, for an $n$-dimensional vector $\mathbf{v} = [v_1, v_2, \ldots, v_n]$:

$$|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$$

This formula assumes that each component of the vector lies in an axis system where all axes are orthogonal to each other (such as x-axis, y-axis, z-axis). We refer to these as **Euclidean vectors**.

### **Vector Equivalence**

**Important Property:** Vectors with the same magnitude and direction are considered identical, regardless of their starting position in space.

For example, three vectors in 2D space are the same vector if they have:
- The same length (magnitude)
- The same direction
- Different starting points do not affect vector equality

### **Linearly Independent Axes**

In any axis system, each axis must be **linearly independent** of the others. This means that no axis can be expressed as a linear combination of the other axes.

### **Three Equivalent Definitions of Linear Independence**

Given axes $x_1, x_2, \ldots, x_n$, the following statements are equivalent:

1. **No axis as linear combination:** Every axis cannot be expressed as a linear combination of all other axes.

2. **Uniqueness of representation:** For any constants $a_1, a_2, \ldots, a_n$ and $b_1, b_2, \ldots, b_n$:
   $$a_1x_1 + a_2x_2 + \cdots + a_nx_n = b_1x_1 + b_2x_2 + \cdots + b_nx_n$$
   if and only if $a_1 = b_1, a_2 = b_2, \ldots, a_n = b_n$

3. **Trivial null combination:** For any constants $a_1, a_2, \ldots, a_n$:
   $$a_1x_1 + a_2x_2 + \cdots + a_nx_n = 0$$
   if and only if $a_1 = 0, a_2 = 0, \ldots, a_n = 0$

### **Constants and Scalars**

A **constant** or **scalar** refers to any real number. Scalars are used to scale vectors (change their magnitude) without affecting their direction.

### **Dimensional Classification**

A vector is fundamentally a **1-dimensional object** regardless of how many components are used to describe it:

- **0-dimensional object:** A single point (no extent)
- **1-dimensional object:** Has length only (like a vector)
- **2-dimensional object:** Has area (like a plane or surface)
- **3-dimensional object:** Has volume (like a solid)

The number of components in a vector (2D, 3D, nD) describes the **space** in which the vector exists, not the **dimensionality** of the vector object itself. The vector remains a 1-dimensional entity with length and direction.
