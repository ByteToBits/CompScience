### **Definitions**

A **Set is a Collection of Objects**. These Objects in a set are called **elements** of the set. For example, as Set of Prime Numbers between 2 to 20 is $P = \{2,3,5,7,11,13,17,19\}$. 

- Objects of a Set has **No Order** and therefore the order in which the elements are describes does not matter. 
- Set can only contain **Unique Objects**. 
- Notation-wise, it is a convention to use capital letters such as A, B, C, to denote the name of a set.
- **{}** to enclose all elements of a set. $P = \{2,3,5,7,11,13,17,19\}$ hence $7 ∈ P$ and $8 ∉ P$
