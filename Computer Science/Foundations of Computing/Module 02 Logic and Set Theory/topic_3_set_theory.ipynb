### **Set Theory Definitions**

A **Set is a Collection of Objects**. These Objects in a set are called **elements** of the set. For example, as Set of Prime Numbers between 2 to 20 is $P = \{2,3,5,7,11,13,17,19\}$. 

| Property | Description | Example |
|----------|-------------|---------|
| **No Order** | The order in which the elements are described does not matter | $\{1, 2, 3\} = \{3, 1, 2\}$ |
| **Unique Objects** | Sets can only contain unique elements (no duplicates) | $\{1, 2, 2, 3\} = \{1, 2, 3\}$ |
| **Notation** | Capital letters (A, B, C) are conventionally used to denote sets | $A = \{1, 2, 3\}$, $B = \{x \in \mathbb{N} \mid x < 5\}$ |
| **Ø or {}** | Elements of a set are enclosed in curly braces or indcated by symbol  Ø means Null | $P = \{\}$ |
| **Representation Forms** | Sets can be described in Enumerated Form or Set Builder Form | Enumerated: $E = \{2, 4, 6, 8\}$<br>Set Builder: $E = \{x \in \mathbb{N} \mid x \text{ is even and } x < 10\}$ |
| **Cardinality** | The number of elements in a set, denoted by $\|A\|$ | If $A = \{a, b, c\}$, then $\|A\| = 3$ |



Cardinality of a Power Set can be written as shown below for A Set of $n$ number of elements: 

$$\text{Cardinality of Power Set (A)} =  2^N$$

**Q1) How many elements are there for the power set of an empty set?**
- a) 2
- b) 1
- c) Infinity
- d) 0
<br>

**Answer: b** 

*Explanation: The power set of an empty set $∅$ contains only one element: the empty set itself. This follows from the formula $2^n$ where n is the number of elements in the original set. Since $n = 0$ for an empty set, we have $2^0 = 1$*

<br>

**Q2) How many elements are there in the power set of set $A = \{1, 2, 3, 4\}$?**
- a) 8
- b) 16
- c) 32
- d) 4
<br>

**Answer: b** 

*Explanation: For a set with $n$ elements, the power set contains $2^n$ elements. Since A has 4 elements, its power set $P(A)$ will have $2^4 = 16$ elements, including the empty set, all singleton sets, all pairs, all triplets, and the set itself.*

<br>

- a) Sub

- d) Super

**Q4) What is the cardinality of the power set of a set with $n$ number of elements?**
- a) $2n + 1$
- b) $2n$
- c) $N$
- d) $2^n$
<br>

**Answer: d** 

*Explanation: If a set has n elements, then its power set (the set of all possible subsets) has exactly $2^n$ elements. This can be proven using combinatorial principles, as each element can either be included or excluded in a subset, giving 2 choices for each of the n elements.*

**Q5) If $A = \{0, 1\}$ and $B = \{1, 2, 3\}$ what is $A \times B$?**

**Answer:** $A \times B = \{(0,1), (0,2), (0,3), (1,1), (1,2), (1,3)\}$

*Explanation: The Cartesian product $A \times B$ is the set of all ordered pairs $(a,b)$ where $a \in A$ and $b \in B$. The symbol "$\times$" represents the Cartesian product operation. We form all possible combinations taking one element from set $A$ as the first component and one element from set $B$ as the second component.*

<br>

**Q6) If $A = \{0, 1\}$ what is $A \times \mathbb{N}$?**

**Answer:** $A \times \mathbb{N} = \{(0,0), (0,1), (0,2), (0,3), ..., (1,0), (1,1), (1,2), (1,3), ...\}$

**Alternate Answer:** $A \times \mathbb{N} = \{(x,y) : x \in A, y \in \mathbb{N}\}$

*Explanation: Here, $\mathbb{N}$ represents the set of natural numbers $\{0, 1, 2, 3, ...\}$. The Cartesian product $A \times \mathbb{N}$ creates an infinite set of ordered pairs where the first element is from set $A$ (either 0 or 1) and the second element is any natural number. This produces two infinite "streams" of ordered pairs, one starting with 0 and one starting with 1.*

*The alternate answer uses set-builder notation, where $(x,y)$ represents an ordered pair, $x \in A$ means x is an element of set A, and $y \in \mathbb{N}$ means y is an element of the natural numbers. This notation describes the same set but in a more compact mathematical form, defining the set by the conditions its elements must satisfy rather than listing them out explicitly. This notation is especially useful for infinite sets like this one that cannot be fully enumerated.*

### **Set Operations**

Below is a comprehensive table of the five fundamental operations on sets, with examples based on the Venn diagram:

| Operation | Symbol | Definition | Set Builder Form | Enumerated Form |
|-----------|--------|------------|--------------|--------------------------|
| **Union** | ∪ | All elements that are in either set A or set B (or both) | A ∪ B = {x : x ∈ A OR x ∈ B} | A ∪ B = {Peter, Thomas, Eric, Lucas, Pedro} |
| **Intersection** | ∩ | All elements that are in both set A and set B | A ∩ B = {x : x ∈ A AND x ∈ B} | A ∩ B = {Eric} |
| **Difference** | − | All elements that are in set A but not in set B | A − B = {x : x ∈ A AND x ∉ B} | A − B = {Peter, Thomas} |
| **Symmetric Difference** | ∆ or ⊖ | All elements that are in either set A or set B, but not in both | A ∆ B = {x : (x ∈ A AND x ∉ B) OR (x ∉ A AND x ∈ B)} | A ∆ B = {Peter, Thomas, Lucas, Pedro} |
| **Complement** | Aᶜ or A' or A̅ | All elements in the universal set that are not in set A | Aᶜ = {x : x ∈ U AND x ∉ A} | Aᶜ = {Lucas, Pedro, Annie, Claire} |

### **Cardinality of Set Operations**

| Operation | Definition | Set Result | Cardinality Formula | Calculation | Result |
|-----------|------------|------------|---------------------|-------------|--------|
| **A** | Elements in set A | {Peter, Thomas, Eric} | n(A) | Count elements in A | 3 |
| **B** | Elements in set B | {Eric, Lucas, Pedro} | n(B) | Count elements in B | 3 |
| **A ∪ B** | Elements in A OR B | {Peter, Thomas, Eric, Lucas, Pedro} | n(A) + n(B) - n(A ∩ B) | 3 + 3 - 1 | 5 |
| **A ∩ B** | Elements in A AND B | {Eric} | n(A) + n(B) - n(A ∪ B) | 3 + 3 - 5 | 1 |
| **A - B** | Elements in A but not in B | {Peter, Thomas} | n(A) - n(A ∩ B) | 3 - 1 | 2 |
| **B - A** | Elements in B but not in A | {Lucas, Pedro} | n(B) - n(A ∩ B) | 3 - 1 | 2 |
| **A ∆ B** | Elements in A or B but not both | {Peter, Thomas, Lucas, Pedro} | n(A ∪ B) - n(A ∩ B) | 5 - 1 | 4 |
| **A^c** | Elements in U but not in A | {Lucas, Pedro, Claire, Annie} | n(U) - n(A) | 7 - 3 | 4 |
| **B^c** | Elements in U but not in B | {Peter, Thomas, Claire, Annie} | n(U) - n(B) | 7 - 3 | 4 |
| **U** | All elements in universe | {Peter, Thomas, Eric, Lucas, Pedro, Claire, Annie} | n(U) | Count elements in U | 7 |

### **Cartesian Product of Sets**

The Cartesian product of sets A and B (denoted as A × B) is the set of all ordered pairs (x,y) where x ∈ A and y ∈ B.

#### **Definition**
- **Formal notation**: A × B = {(x,y) : x ∈ A, y ∈ B}
- **Cardinality**: n(A × B) = n(A) × n(B)

#### **Example**
If A = {1, 2, 3} and B = {3, 4, 5}, then:

A × B = {(1,3), (1,4), (1,5), (2,3), (2,4), (2,5), (3,3), (3,4), (3,5)}

Here:
- n(A) = 3
- n(B) = 3
- n(A × B) = 3 × 3 = 9

### **Law on Sets**

| Law | Description | Union Operation | Intersection Operation |
|-----|-------------|----------------|------------------------|
| **Associative Law** | The grouping of operations doesn't affect the result | $(A \cup B) \cup C = A \cup (B \cup C)$ | $(A \cap B) \cap C = A \cap (B \cap C)$ |
| **Identity Law** | There exists a set that leaves other sets unchanged when combined | $A \cup \emptyset = A$ | $A \cap U = A$ |
| **Idempotent Law** | Combining a set with itself yields the original set | $A \cup A = A$ | $A \cap A = A$ |
| **Commutative Law** | The order of sets doesn't matter when performing operations | $A \cup B = B \cup A$ | $A \cap B = B \cap A$ |
| **Domination Law** | There exists a set that dominates in each operation | $A \cup U = U$ | $A \cap \emptyset = \emptyset$ |
| **Distributive Law** | One operation distributes over the other | $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$ | $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$ |

**Q1) Let,**

**$A = \{x : x \text{ is a natural number and a factor of } 18\}$ and**
**$B = \{x : x \text{ is a natural number and less than } 6\}$.**
**Find $A \cap B$.**

- a) $\{1, 2, 3\}$
- b) $\{1, 2, 3, 6, 9\}$
- c) $\{1, 2, 3, 4, 5, 6, 9\}$
- d) $\{1, 2, 3, 4, 5\}$

**Answer: a) $\{1, 2, 3\}$** 

*Explanation: Set $A$ contains all factors of 18, which are $\{1, 2, 3, 6, 9, 18\}$. Set $B$ contains natural numbers less than 6, which are $\{1, 2, 3, 4, 5\}$. The intersection $A \cap B$ consists of elements that belong to both sets, which are $\{1, 2, 3\}$.*

<br>

**Q2) Given,**

**$A = \{2, 3, 4, 5, 6\}$ and $B = \{3, 5, 7, 8, 9\}$,**
**find the symmetric difference ($\Delta$) between $A$ and $B$.**

- a) $\{3, 5\}$
- b) $\{2, 4, 6, 7, 8, 9\}$
- c) $\{2, 3, 4, 5, 6, 7, 8, 9\}$
- d) $\{2, 3, 4, 5, 6\}$

**Answer: b) $\{2, 4, 6, 7, 8, 9\}$** 

*Explanation: The symmetric difference of sets $A$ and $B$, denoted as $A \Delta B$, is defined as $(A \cup B) - (A \cap B)$, which consists of elements that are in either $A$ or $B$ but not in both. First, $A \cap B = \{3, 5\}$. Then, $A \Delta B = (A - B) \cup (B - A) = \{2, 4, 6\} \cup \{7, 8, 9\} = \{2, 4, 6, 7, 8, 9\}$.*

<br>
