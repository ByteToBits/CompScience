### **Introduction to Derivatives**

Derivatives measure the **rate of change** of a function with respect to its variable. They are fundamental tools in calculus for analyzing function behavior.

#### **Key Notation:**
- **First derivative**: $f'(x)$, $\frac{dy}{dx}$, or $\frac{df}{dx}$
- **Second derivative**: $f''(x)$ or $\frac{d^2y}{dx^2}$

---

### **First-Order Derivatives**

#### **Definition and Purpose:**
The first derivative represents the **instantaneous rate of change** of a function at any given point.

#### **Applications:**
- **Slope of tangent lines** at specific points
- **Velocity** in physics (rate of change of position)
- **Current analysis** in electrical circuits
- **Optimization** problems (finding maxima and minima)

#### **Mathematical Interpretation:**
- $f'(x) > 0$: Function is **increasing**
- $f'(x) < 0$: Function is **decreasing**
- $f'(x) = 0$: Function has a **critical point**

---

### **Second-Order Derivatives**

#### **Definition and Purpose:**
The second derivative measures the **rate of change of the rate of change** - essentially how the slope itself is changing.

#### **Concavity Analysis:**
- $f''(x) > 0$: **Concave up** (convex shape) - curve opens upward
- $f''(x) < 0$: **Concave down** - curve opens downward
- $f''(x) = 0$: Possible **inflection point**

#### **Visual Understanding:**
- **Concave up**: Tangent line lies below the curve
- **Concave down**: Tangent line lies above the curve



---

### **Practical Example: Analyzing f(x) = x + 1**

Let's work through a complete analysis of the parabola $f(x) = x^2 + 1$.

**Step 1: Find the First Derivative**
$$f'(x) = \frac{d}{dx}(x^2 + 1) = 2x$$

**Step 2: Find the Second Derivative**
$$f''(x) = \frac{d}{dx}(2x) = 2$$

**Step 3: Analysis at x = -2**
- **Function value**: $f(-2) = (-2)^2 + 1 = 5$
- **First derivative**: $f'(-2) = 2(-2) = -4$
- **Second derivative**: $f''(-2) = 2 > 0$ (concave up)



---

### **Finding the Tangent Line at x = -2**

**Method 1: Using the Derivative (Recommended)**
The slope of the tangent line at any point is given by the first derivative.

**Point of tangency**: $(-2, 5)$
**Slope**: $m = f'(-2) = -4$

Using point-slope form: $y - y_1 = m(x - x_1)$
$$y - 5 = -4(x - (-2))$$
$$y - 5 = -4(x + 2)$$
$$y = -4x - 8 + 5$$
$$y = -4x - 3$$

**Method 2: Using Two Points (Less Efficient)**
If we identify points $(-1, 1)$ and $(-2, 5)$ on the tangent line:
$$\text{Slope} = \frac{5 - 1}{-2 - (-1)} = \frac{4}{-1} = -4$$

Then using point-slope form with $(-2, 5)$:
$$y - 5 = -4(x + 2) \Rightarrow y = -4x - 3$$

**Derivative of the Tangent Line**
Since the tangent line is $y = -4x - 3$:
$$\frac{dy}{dx} = -4$$

This confirms our slope calculation.

### **Partial Derivatives**

**Partial derivatives** calculate the rate of change of a multivariable function with respect to one variable while **holding all other variables constant**.

#### **Key Concepts:**
- **Purpose**: Analyze how a function changes when only one variable varies
- **Applications**: Optimization problems, surface analysis, finding critical points
- **Notation**: $\frac{\partial f}{\partial x}$, $f_x$, or $f'_x(x,y)$

#### **The Partial Derivative Symbol: **
The symbol $\partial$ (partial) is called:
- "del" 
- "dee"
- "curly dee"
- "dabba"

#### **Basic Examples**

#### **Single Variable vs. Multivariable:**
- **Single variable**: If $f(x) = x^3$, then $f'(x) = 3x^2$
- **Multivariable**: If $f(x,y) = x^3 + y^2$, then:
  - $\frac{\partial f}{\partial x} = 3x^2$ (treat $y$ as constant)
  - $\frac{\partial f}{\partial y} = 2y$ (treat $x$ as constant)

#### **Three-Variable Example:**
Given $f(x,y,z) = z^3 - x^2y + xyz$

**Step-by-step calculation:**
- $\frac{\partial f}{\partial x} = \frac{\partial}{\partial x}(z^3 - x^2y + xyz) = -2xy + yz$
- $\frac{\partial f}{\partial y} = \frac{\partial}{\partial y}(z^3 - x^2y + xyz) = -x^2 + xz$  
- $\frac{\partial f}{\partial z} = \frac{\partial}{\partial z}(z^3 - x^2y + xyz) = 3z^2 + xy$

---

### **Partial Derivative Rules**

#### **Product Rule for Partial Derivatives**

| **Function** | **Rule** |
|-------------|----------|
| $u = f(x,y) \cdot g(x,y)$ | $\frac{\partial u}{\partial x} = g(x,y)\frac{\partial f}{\partial x} + f(x,y)\frac{\partial g}{\partial x}$ |
| | $\frac{\partial u}{\partial y} = g(x,y)\frac{\partial f}{\partial y} + f(x,y)\frac{\partial g}{\partial y}$ |

#### **Quotient Rule for Partial Derivatives**

| **Function** | **Rule** |
|-------------|----------|
| $u = \frac{f(x,y)}{g(x,y)}$ where $g(x,y) \neq 0$ | $\frac{\partial u}{\partial x} = \frac{g(x,y)\frac{\partial f}{\partial x} - f(x,y)\frac{\partial g}{\partial x}}{[g(x,y)]^2}$ |
| | $\frac{\partial u}{\partial y} = \frac{g(x,y)\frac{\partial f}{\partial y} - f(x,y)\frac{\partial g}{\partial y}}{[g(x,y)]^2}$ |

#### **Power Rule for Partial Derivatives**

| **Function** | **Rule** |
|-------------|----------|
| $u = [f(x,y)]^n$ | $\frac{\partial u}{\partial x} = n[f(x,y)]^{n-1}\frac{\partial f}{\partial x}$ |
| | $\frac{\partial u}{\partial y} = n[f(x,y)]^{n-1}\frac{\partial f}{\partial y}$ |

#### **Chain Rule for Partial Derivatives**

#### **Simple Chain Rule:**
| **Function** | **Rule** |
|-------------|----------|
| $y = f(u)$ and $u = g(x)$ | $\frac{d}{dx}[f(g(x))] = f'(g(x))g'(x)$ |
| i.e., $y = f(g(x))$ | |

#### **Extended Chain Rule:**
| **Function** | **Rule** |
|-------------|----------|
| $y = f(g(h(x)))$ | $\frac{d}{dx}[f(g(h(x)))] = f'(g(h(x)))g'(h(x))h'(x)$ |

#### **Chain Rule with Two Independent Variables:**
| **Function** | **Rule** |
|-------------|----------|
| $x = g(u,v)$, $y = h(u,v)$ | $\frac{\partial z}{\partial u} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial u} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial u}$ |
| and $z = f(x,y)$ | $\frac{\partial z}{\partial v} = \frac{\partial z}{\partial x}\frac{\partial x}{\partial v} + \frac{\partial z}{\partial y}\frac{\partial y}{\partial v}$ |
| i.e., $z = f(g(u,v), h(u,v))$ | |

#### **Constant Rule**
| **Function** | **Rule** |
|-------------|----------|
| $f(x,y) = c$ (constant) | $\frac{\partial f}{\partial x} = 0$, $\frac{\partial f}{\partial y} = 0$ |

#### **Sum/Difference Rule**
| **Function** | **Rule** |
|-------------|----------|
| $f(x,y) = u(x,y)  v(x,y)$ | $\frac{\partial f}{\partial x} = \frac{\partial u}{\partial x}  \frac{\partial v}{\partial x}$ |
| | $\frac{\partial f}{\partial y} = \frac{\partial u}{\partial y}  \frac{\partial v}{\partial y}$ |

#### **Constant Multiple Rule**
| **Function** | **Rule** |
|-------------|----------|
| $f(x,y) = c \cdot u(x,y)$ | $\frac{\partial f}{\partial x} = c \cdot \frac{\partial u}{\partial x}$ |
| | $\frac{\partial f}{\partial y} = c \cdot \frac{\partial u}{\partial y}$ |

---

### **Higher-Order Partial Derivatives**

#### **Second-Order Partial Derivatives:**
For $f(x,y)$:
- $\frac{\partial^2 f}{\partial x^2} = f_{xx}$ (pure second partial)
- $\frac{\partial^2 f}{\partial y^2} = f_{yy}$ (pure second partial)
- $\frac{\partial^2 f}{\partial x \partial y} = f_{xy}$ (mixed partial)
- $\frac{\partial^2 f}{\partial y \partial x} = f_{yx}$ (mixed partial)

#### **Clairaut's Theorem:**
If $f_{xy}$ and $f_{yx}$ are continuous, then $f_{xy} = f_{yx}$

---

### **Practical Examples**

### **Example 1: Product Rule Application**
Given $f(x,y) = (x^2 + y)(3x - y^2)$

Let $u = x^2 + y$ and $v = 3x - y^2$

$\frac{\partial f}{\partial x} = v\frac{\partial u}{\partial x} + u\frac{\partial v}{\partial x} = (3x - y^2)(2x) + (x^2 + y)(3) = 6x^2 - 2xy^2 + 3x^2 + 3y = 9x^2 - 2xy^2 + 3y$

### **Example 2: Chain Rule Application**
Given $z = \sin(x^2 + y^2)$

Let $u = x^2 + y^2$, so $z = \sin(u)$

$\frac{\partial z}{\partial x} = \cos(u) \cdot \frac{\partial u}{\partial x} = \cos(x^2 + y^2) \cdot 2x = 2x\cos(x^2 + y^2)$

$$y - 13 = 22(x + 1)$$

**Q1) Find the partial derivatives of $f(x, y) = x^2y + \sin x + \cos y$**

*Explanation: To find partial derivatives, we differentiate with respect to one variable while treating other variables as constants.*

$$\text{Given: } f(x, y) = x^2y + \sin x + \cos y$$

**Step 1: Find the partial derivative with respect to x ($f_x'$):**

$$f_x'(x, y) = \frac{\partial}{\partial x}(x^2y + \sin x + \cos y)$$

$$f_x'(x, y) = \frac{\partial}{\partial x}(x^2y) + \frac{\partial}{\partial x}(\sin x) + \frac{\partial}{\partial x}(\cos y)$$

$$f_x'(x, y) = 2xy + \cos x + 0$$

$$f_x'(x, y) = 2xy + \cos x$$

**Step 2: Find the partial derivative with respect to y ($f_y'$):**

$$f_y'(x, y) = \frac{\partial}{\partial y}(x^2y + \sin x + \cos y)$$

$$f_y'(x, y) = \frac{\partial}{\partial y}(x^2y) + \frac{\partial}{\partial y}(\sin x) + \frac{\partial}{\partial y}(\cos y)$$

$$f_y'(x, y) = x^2 + 0 + (-\sin y)$$

$$f_y'(x, y) = x^2 - \sin y$$

**Q5) Find the partial derivatives of $f(x, y) = \sin(x)y$**

**Answer: a** <br>
*Explanation: This is a product of a trigonometric function and a variable. We apply the chain rule for the sine function.*

$$\text{Given: } f(x, y) = \sin(x)y$$

**Step 1: Find the partial derivative with respect to x ($f_x'$):**

$$f_x'(x, y) = \frac{\partial}{\partial x}(\sin(x)y)$$

$$\text{Since } y \text{ is treated as a constant with respect to } x:$$

$$f_x'(x, y) = y \cdot \frac{\partial}{\partial x}(\sin(x))$$

$$f_x'(x, y) = y \cdot \cos(x)$$

$$f_x'(x, y) = \cos(x)y$$

**Step 2: Find the partial derivative with respect to y ($f_y'$):**

$$f_y'(x, y) = \frac{\partial}{\partial y}(\sin(x)y)$$

$$\text{Since } \sin(x) \text{ is treated as a constant with respect to } y:$$

$$f_y'(x, y) = \sin(x) \cdot \frac{\partial}{\partial y}(y)$$

$$f_y'(x, y) = \sin(x) \cdot 1$$

$$f_y'(x, y) = \sin(x)$$
