Integration is the reverse of differentiation. If you have a derivative $f'(x)$, integration helps you find the original function $f(x)$. The symbol for integration is $\int$. If differentiation breaks down a function, integration builds it back up.

### **Two Main Types of Integrals**

#### 1. Indefinite Integrals
- **Form:** $\int f(x) \, dx = F(x) + C$
- **Key feature:** Always includes a constant $C$
- **Why the constant?** Because many functions can have the same derivative

**Example:** If $f'(x) = 2x$, then $f(x)$ could be $x^2$, $x^2 + 5$, or $x^2 + 100$. So we write: $\int 2x \, dx = x^2 + C$

#### 2. Definite Integrals
- **Form:** $\int_a^b f(x) \, dx = F(b) - F(a)$
- **Key feature:** Gives a specific number (no constant $C$)
- **Use:** Calculates area under a curve between two points

### **Essential Integration Formulas**

#### **Basic Functions**
| Function | Integral |
|----------|----------|
| $\int 1 \, dx$ | $x + C$ |
| $\int x^n \, dx$ | $\frac{x^{n+1}}{n+1} + C$ |
| $\int \frac{1}{x} \, dx$ | $\ln|x| + C$ |

#### **Trigonometric Functions**
| Function | Integral |
|----------|----------|
| $\int \sin x \, dx$ | $-\cos x + C$ |
| $\int \cos x \, dx$ | $\sin x + C$ |
| $\int \sec^2 x \, dx$ | $\tan x + C$ |

#### **Exponential Functions**
| Function | Integral |
|----------|----------|
| $\int e^x \, dx$ | $e^x + C$ |

### **Integration Rules**

1. **Sum Rule:** $\int (f + g) \, dx = \int f \, dx + \int g \, dx$
2. **Constant Rule:** $\int c \cdot f(x) \, dx = c \int f(x) \, dx$
3. **Difference Rule:** $\int (f - g) \, dx = \int f \, dx - \int g \, dx$

### **Example 1:**

**Given:** $f''(t) = \cos t$, $f'(0) = 3$, and $f(0) = 5$. Find $f(t)$.

**Step 1:** Find $f'(t)$
$$f'(t) = \int \cos t \, dt = \sin t + C_1$$

**Step 2:** Use $f'(0) = 3$
$$\sin(0) + C_1 = 3 \rightarrow C_1 = 3$$
$$f'(t) = \sin t + 3$$

**Step 3:** Find $f(t)$
$$f(t) = \int (\sin t + 3) \, dt = -\cos t + 3t + C_2$$

**Step 4:** Use $f(0) = 5$
$$-\cos(0) + 3(0) + C_2 = 5 \rightarrow C_2 = 6$$

**Answer:** $f(t) = -\cos t + 3t + 6$

**Q1) Calculate the following integral: $\int_0^3 |x - 1| dx$**

- a) 0
- b) 2
- c) 3/2
- d) 5/2

<br>

**Answer: d** <br>
*Explanation: Since we have an absolute value function, we need to consider where the expression inside changes sign and split the integral accordingly.*

$\text{Given: } \int_0^3 |x - 1| dx$

**Step 1: Determine where $x - 1 = 0$:**

$x - 1 = 0 \Rightarrow x = 1$

**Step 2: Analyze the sign of $(x - 1)$ on different intervals:**

$\text{For } x \in [0, 1]: x - 1 < 0, \text{ so } |x - 1| = -(x - 1) = 1 - x$

$\text{For } x \in [1, 3]: x - 1 > 0, \text{ so } |x - 1| = x - 1$

**Step 3: Split the integral at $x = 1$:**

$\int_0^3 |x - 1| dx = \int_0^1 |x - 1| dx + \int_1^3 |x - 1| dx$

$= \int_0^1 (1 - x) dx + \int_1^3 (x - 1) dx$

**Step 4: Evaluate the first integral:**

$\int_0^1 (1 - x) dx = \left[x - \frac{x^2}{2}\right]_0^1$

$= \left(1 - \frac{1}{2}\right) - \left(0 - 0\right) = \frac{1}{2}$

**Step 5: Evaluate the second integral:**

$\int_1^3 (x - 1) dx = \left[\frac{x^2}{2} - x\right]_1^3$

$= \left(\frac{9}{2} - 3\right) - \left(\frac{1}{2} - 1\right)$

$= \frac{9}{2} - 3 - \frac{1}{2} + 1 = \frac{8}{2} - 2 = 4 - 2 = 2$

**Step 6: Add both parts:**

$\int_0^3 |x - 1| dx = \frac{1}{2} + 2 = \frac{5}{2}$

**Q2) Find the following integral: $\int \frac{4x^2 + 1}{x^2} dx, x \neq 0$**

- a) $x^2 - x + C$
- b) $2x^2 - x - 1 + C$
- c) $2x - x - 1 + C$
- d) $2x^2 - (1/x) + C$

<br>

**Answer: d** <br>
*Explanation: We need to split the fraction and integrate each term separately.*

$\text{Given: } \int \frac{4x^2 + 1}{x^2} dx$

**Step 1: Split the fraction:**

$\frac{4x^2 + 1}{x^2} = \frac{4x^2}{x^2} + \frac{1}{x^2}$

$= 4 + \frac{1}{x^2}$

$= 4 + x^{-2}$

**Step 2: Rewrite the integral:**

$\int \frac{4x^2 + 1}{x^2} dx = \int \left(4 + x^{-2}\right) dx$

$= \int 4 dx + \int x^{-2} dx$

**Step 3: Integrate each term:**

$\int 4 dx = 4x$

$\int x^{-2} dx = \int x^{-2} dx = \frac{x^{-2+1}}{-2+1} = \frac{x^{-1}}{-1} = -x^{-1} = -\frac{1}{x}$

**Step 4: Combine the results:**

$\int \frac{4x^2 + 1}{x^2} dx = 4x - \frac{1}{x} + C$

$= 4x - \frac{1}{x} + C$

*Note: This can also be written as $4x - x^{-1} + C$. Looking at the options, the closest match considering the coefficient is $2x^2 - (1/x) + C$, but there appears to be a discrepancy in the coefficient. The correct answer should be $4x - (1/x) + C$.*

**Q1) Find the integral: $\int (1/x^7) dx$**

**Answer:** $-\frac{1}{6x^6} + C$ <br>
*Explanation: This is an integration of a power function with a negative exponent.*

$\text{Given: } \int \frac{1}{x^7} dx$

**Step 1: Rewrite using negative exponent:**

$\int \frac{1}{x^7} dx = \int x^{-7} dx$

**Step 2: Apply the power rule for integration:**

$\text{Power rule: } \int x^n dx = \frac{x^{n+1}}{n+1} + C \text{ (where } n \neq -1\text{)}$

$\int x^{-7} dx = \frac{x^{-7+1}}{-7+1} + C$

$= \frac{x^{-6}}{-6} + C$

$= -\frac{1}{6}x^{-6} = \frac{-1}{6x^6}$

**Q2) Find the integral: $\int (1/\sin^2 x) dx$**

**Answer:** $-\cot x + C$ <br>
*Explanation: This integral involves the cosecant squared function, and can also be solved using an alternative approach with substitution.*

$\text{Given: } \int \frac{1}{\sin^2 x} dx$

**Step 1: Method 1 - Direct recognition:**

$\frac{1}{\sin^2 x} = \csc^2 x$

$\int \csc^2 x \, dx = -\cot x + C$

**Step 2: Method 2 - Using substitution (as shown in the provided solution):**

$\text{Assume } u = \cot x = \frac{\cos x}{\sin x}$

$du = \frac{-\sin x \cdot \sin x - \cos x \cdot \cos x}{\sin^2 x} dx$

$= \frac{-(\sin^2 x + \cos^2 x)}{\sin^2 x} dx$

$= \frac{-1}{\sin^2 x} dx = -\csc^2 x \, dx$

**Step 3: Therefore:**

$\int \frac{1}{\sin^2 x} dx = \int \csc^2 x \, dx = -\cot x + C$


<br><br>

**Q3) Find the integral: $\int \frac{\cos x}{\sin^2 x} dx$**

**Answer:** $-\frac{1}{\sin x} + C$ or $-\csc x + C$ <br>
*Explanation: This integral can be solved using substitution method.*

$\text{Given: } \int \frac{\cos x}{\sin^2 x} dx$

**Step 1: Use substitution method:**

$\text{Let } u = \sin x$

$\text{Then } du = \cos x \, dx$

**Step 2: Substitute into the integral:**

$\int \frac{\cos x}{\sin^2 x} dx = \int \frac{1}{\sin^2 x} \cos x \, dx$

$= \int \frac{1}{u^2} du$

$= \int u^{-2} du$

**Step 3: Apply the power rule:**

$\int u^{-2} du = \frac{u^{-2+1}}{-2+1} + C$

$= \frac{u^{-1}}{-1} + C$

$= -u^{-1} + C$

**Step 4: Substitute back:**

$= -(\sin x)^{-1} + C$

$= -\frac{1}{\sin x} + C = -\csc x + C$

**Q4) Find the integral: $\int (1/x) dx$**

**Answer:** $\ln|x| + C$ <br>
*Explanation: This is the standard integral of the reciprocal function.*

$$\text{Given: } \int \frac{1}{x} dx$$

**Step 1: Apply the standard formula:**

$$\text{Standard formula: } \int \frac{1}{x} dx = \ln|x| + C$$

$$\text{Therefore: } \int \frac{1}{x} dx = \ln|x| + C$$

*Note: The absolute value is important because the natural logarithm is only defined for positive arguments, but we want our antiderivative to work for both positive and negative values of x (where x  0).*

**Q5) Find the integral: $\int (1-x^2)^{-1/2} dx$**

**Answer:** $\sin^{-1}(x) + C$ or $\arcsin(x) + C$ <br>
*Explanation: This integral can be solved using trigonometric substitution.*

$\text{Given: } \int (1-x^2)^{-1/2} dx$

**Step 1: Rewrite the integrand:**

$\int (1-x^2)^{-1/2} dx = \int \frac{1}{\sqrt{1-x^2}} dx$

**Step 2: Use trigonometric substitution (as shown in the provided solution):**

$\text{Assume } x = \sin u$

$\text{Then } dx = \cos u \, du$

**Step 3: Substitute:**

$\int (1-x^2)^{-1/2} dx = \int (1-\sin^2 u)^{-1/2} \cos u \, du$

$= \int (\cos^2 u)^{-1/2} \cos u \, du$

$= \int \frac{1}{\sqrt{\cos^2 u}} \cdot \cos u \, du$

$= \int \frac{1}{|\cos u|} \cdot \cos u \, du$

**Step 4: For the principal domain where $\cos u > 0$:**

$= \int \frac{\cos u}{\cos u} du = \int 1 \, du = u$

**Step 5: Substitute back:**

$= \sin^{-1}(x) + C$

**Q6) Find the definite integral: $\int_0^3 x^2 dx$**

**Answer:** $9$ <br>
*Explanation: This is a definite integral of a polynomial function.*

$$\text{Given: } \int_0^3 x^2 dx$$

**Step 1: Find the antiderivative:**

$$\int x^2 dx = \frac{x^{2+1}}{2+1} + C = \frac{x^3}{3} + C$$

**Step 2: Apply the Fundamental Theorem of Calculus:**

$$\int_0^3 x^2 dx = \left[\frac{x^3}{3}\right]_0^3$$

**Step 3: Evaluate at the limits:**

$$= \frac{(3)^3}{3} - \frac{(0)^3}{3}$$

$$= \frac{27}{3} - \frac{0}{3}$$

$$= 9 - 0 = 9$$

**Q7) Find the definite integral: $\int_0^{\pi/4} \sin(2x) dx$**

**Answer:** $\frac{1}{2}$ <br>
*Explanation: This involves integrating a sine function with a coefficient inside the argument.*

$$\text{Given: } \int_0^{\pi/4} \sin(2x) dx$$

**Step 1: Find the antiderivative using substitution:**

$$\text{Let } u = 2x, \text{ then } du = 2dx, \text{ so } dx = \frac{du}{2}$$

$$\int \sin(2x) dx = \int \sin(u) \cdot \frac{du}{2} = \frac{1}{2} \int \sin(u) du$$

$$= \frac{1}{2}(-\cos(u)) + C = -\frac{1}{2}\cos(2x) + C$$

**Step 2: Apply the Fundamental Theorem of Calculus:**

$$\int_0^{\pi/4} \sin(2x) dx = \left[-\frac{1}{2}\cos(2x)\right]_0^{\pi/4}$$

**Step 3: Evaluate at the limits:**

$$= -\frac{1}{2}\cos\left(2 \cdot \frac{\pi}{4}\right) - \left(-\frac{1}{2}\cos(2 \cdot 0)\right)$$

$$= -\frac{1}{2}\cos\left(\frac{\pi}{2}\right) + \frac{1}{2}\cos(0)$$

$$= -\frac{1}{2}(0) + \frac{1}{2}(1)$$

$$= 0 + \frac{1}{2} = \frac{1}{2}$$
