### **Integration: Finding an Area Under a Curve**

Integrals calculate the area under a curve by breaking it into thin rectangles and summing their areas.



### The Rectangle Method
- Split the area under a curve into many thin vertical rectangles
- Each rectangle has:
  - **Width:** Small change in $x$
  - **Height:** Function value $f(x)$
- Sum all rectangle areas to get the total area

### Mathematical Formula
For area between curve $f(x)$ and x-axis from $x = a$ to $x = b$:
$$\text{Area} = \int_a^b f(x) \, dx$$

### Example: $f(x) = x^2$ from $x = -1$ to $x = 2$

**Solution:**
$$\text{Area} = \int_{-1}^{2} x^2 \, dx = \left[\frac{x^3}{3}\right]_{-1}^{2}$$

$$= \frac{2^3}{3} - \frac{(-1)^3}{3} = \frac{8}{3} - \frac{-1}{3} = \frac{8}{3} + \frac{1}{3} = 3$$

**Answer:** 3 square units

### **Integration:Finding Arc Length of a Curve**

Arc length is the distance along a curve between two points. Unlike straight-line distance, it follows the actual path of the curve.

To find the length of a smooth curve, we:
1. Divide the curve into infinitesimally small sections
2. Calculate the length of each small section using Pythagoras' theorem
3. Add up all the small lengths using integration



### **Pythagorean Approach**

For each tiny triangle along the curve:
- **Horizontal leg:** $\Delta x_i$
- **Vertical leg:** $\Delta y_i$ 
- **Hypotenuse:** $\sqrt{\Delta x_i^2 + \Delta y_i^2}$

### **Arc Length Formula Development**

**Step 1:** Sum of all small segments
$$\text{Arc Length} = \int_a^b \sqrt{\Delta x_i^2 + \Delta y_i^2}$$

**Step 2:** Convert to differential form
$$\text{Arc Length} = \int_a^b \sqrt{dx^2 + dy^2}$$

**Step 3:** Factor out $dx$
$$\text{Arc Length} = \int_a^b \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

**Step 4:** Final formula (since $f'(x) = \frac{dy}{dx}$)
$$s = \int_a^b \sqrt{1 + [f'(x)]^2} \, dx$$

### **Worked Example**

**Find the arc length of $f(x) = \frac{x^2}{2} - \frac{\ln(x)}{4}$ from $x = 2$ to $x = 4$**

**Step 1:** Find the derivative
$$f'(x) = \frac{d}{dx}\left(\frac{x^2}{2} - \frac{\ln(x)}{4}\right) = x - \frac{1}{4x}$$

**Step 2:** Calculate $[f'(x)]^2$
$$[f'(x)]^2 = \left(x - \frac{1}{4x}\right)^2 = x^2 - \frac{1}{2} + \frac{1}{16x^2}$$

**Step 3:** Simplify $1 + [f'(x)]^2$
$$1 + [f'(x)]^2 = 1 + x^2 - \frac{1}{2} + \frac{1}{16x^2} = x^2 + \frac{1}{2} + \frac{1}{16x^2}$$

**Step 4:** Factor the expression
$$1 + [f'(x)]^2 = \left(x + \frac{1}{4x}\right)^2$$

**Step 5:** Set up the integral
$$s = \int_2^4 \sqrt{\left(x + \frac{1}{4x}\right)^2} \, dx = \int_2^4 \left(x + \frac{1}{4x}\right) dx$$

**Step 6:** Integrate
$$s = \int_2^4 x \, dx + \int_2^4 \frac{1}{4x} \, dx = \left[\frac{x^2}{2} + \frac{\ln x}{4}\right]_2^4$$

**Step 7:** Evaluate
$$s = \left(\frac{16}{2} + \frac{\ln 4}{4}\right) - \left(\frac{4}{2} + \frac{\ln 2}{4}\right)$$
$$s = 8 + \frac{\ln 4}{4} - 2 - \frac{\ln 2}{4} = 6 + \frac{\ln 4 - \ln 2}{4}$$
$$s = 6 + \frac{\ln 2}{4}$$

**Answer:** The arc length is $6 + \frac{\ln 2}{4}$ units.

### **Integration: Finding the Area of a Surface**

Surface area is the total area of the outer layer of any object. For curved surfaces, calculating this area is more complex than for standard shapes like cylinders, spheres, or cubes. Finding surface area extends the concept of arc length from one dimension (curves) to two dimensions (surfaces).

To find the area of a curved surface:

1. **Divide** the surface into small subregions
2. **Map** each subregion onto a tangent plane
3. **Calculate** the area of each small piece
4. **Sum** all the small areas using double integrals

#### **Understanding the Process**

For each small subregion (the $i$-th piece):
- **Dimensions:** $\Delta x_i$ and $\Delta y_i$
- **Area:** $\Delta A_i = \Delta x_i \cdot \Delta y_i$

Since we're dealing with a surface, we need to consider approximations in **both** the $x$ and $y$ directions, which is why we use double integrals ($\iint$).

#### **Surface Area Formula**

For a surface defined by $z = f(x,y)$, where the partial derivatives $f_x$ and $f_y$ are continuous over a closed, bounded region $R$:

$$S = \iint_R dS = \iint_R \sqrt{1 + f_x(x,y)^2 + f_y(x,y)^2} \, dA$$

#### **Formula Components**

- **$S$** = Total surface area
- **$f_x(x,y)$** = Partial derivative with respect to $x$
- **$f_y(x,y)$** = Partial derivative with respect to $y$
- **$R$** = The region in the $xy$-plane
- **$dA$** = Small area element in the $xy$-plane

#### **Connection to Arc Length**

| **Arc Length (1D)** | **Surface Area (2D)** |
|---------------------|------------------------|
| $s = \int_a^b \sqrt{1 + [f'(x)]^2} \, dx$ | $S = \iint_R \sqrt{1 + f_x^2 + f_y^2} \, dA$ |
| Uses single integral | Uses double integral |
| One derivative $f'(x)$ | Two partial derivatives $f_x, f_y$ |

### **Worksheet Question**

**Q1:** Find the area under the curve $f(x) = x(x-2)(x+2)$ for $-1 \leq x \leq 1$.

*Explanation: To find the area under the curve, we need to evaluate the definite integral. Since the function is negative in the given interval, we'll need to take the absolute value to get the actual area.*

$$\text{Given: } f(x) = x(x-2)(x+2) \text{ for } -1 \leq x \leq 1$$

**Step 1: Expand the function:**

$$f(x) = x(x-2)(x+2)$$

$$= x[(x-2)(x+2)]$$

$$= x[x^2 - 4]$$

$$= x^3 - 4x$$

**Step 2: Analyze the function's behavior in the interval $[-1, 1]$:**

$$\text{At } x = -1: f(-1) = (-1)^3 - 4(-1) = -1 + 4 = 3$$

$$\text{At } x = 0: f(0) = 0^3 - 4(0) = 0$$

$$\text{At } x = 1: f(1) = 1^3 - 4(1) = 1 - 4 = -3$$

**Step 3: Check if the function changes sign in the interval:**

From the graph and our calculations, we can see that:
- $f(x) > 0$ for $x \in [-1, 0)$
- $f(x) < 0$ for $x \in (0, 1]$

**Step 4: Set up the integral for the area:**

Since the function changes sign at $x = 0$, we need to split the integral:

$$\text{Area} = \int_{-1}^{0} |f(x)| dx + \int_{0}^{1} |f(x)| dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx + \int_{0}^{1} |x^3 - 4x| dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx + \int_{0}^{1} -(x^3 - 4x) dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx - \int_{0}^{1} (x^3 - 4x) dx$$

**Step 5: Find the antiderivative:**

$$\int (x^3 - 4x) dx = \frac{x^4}{4} - 4 \cdot \frac{x^2}{2} = \frac{x^4}{4} - 2x^2$$

**Step 6: Evaluate the first integral:**

$$\int_{-1}^{0} (x^3 - 4x) dx = \left[\frac{x^4}{4} - 2x^2\right]_{-1}^{0}$$

$$= \left(\frac{0^4}{4} - 2(0)^2\right) - \left(\frac{(-1)^4}{4} - 2(-1)^2\right)$$

$$= 0 - \left(\frac{1}{4} - 2\right)$$

$$= 0 - \left(\frac{1}{4} - \frac{8}{4}\right)$$

$$= 0 - \left(-\frac{7}{4}\right) = \frac{7}{4}$$

**Step 7: Evaluate the second integral:**

$$\int_{0}^{1} (x^3 - 4x) dx = \left[\frac{x^4}{4} - 2x^2\right]_{0}^{1}$$

$$= \left(\frac{1^4}{4} - 2(1)^2\right) - \left(\frac{0^4}{4} - 2(0)^2\right)$$

$$= \left(\frac{1}{4} - 2\right) - 0$$

$$= \frac{1}{4} - \frac{8}{4} = -\frac{7}{4}$$

**Step 8: Calculate the total area:**

$$\text{Area} = \int_{-1}^{0} (x^3 - 4x) dx - \int_{0}^{1} (x^3 - 4x) dx$$

$$= \frac{7}{4} - \left(-\frac{7}{4}\right)$$

$$= \frac{7}{4} + \frac{7}{4}$$

$$= \frac{14}{4} = \frac{7}{2}$$

**Final Answer:** The area under the curve is $\frac{7}{2}$ square units.

**Q2:** Calculate the arc length of the graph of $f(x) = \frac{4}{3}x^{3/2}$ over the interval $[0, 1]$.

*Explanation: To find the arc length of a curve, we use the arc length formula for a function y = f(x).*

$$\text{Given: } f(x) = \frac{4}{3}x^{3/2} \text{ over the interval } [0, 1]$$

**Step 1: Recall the arc length formula:**

For a function $y = f(x)$ over the interval $[a, b]$, the arc length is:

$$L = \int_a^b \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

**Step 2: Find the derivative of f(x):**

$$f(x) = \frac{4}{3}x^{3/2}$$

$$f'(x) = \frac{d}{dx}\left(\frac{4}{3}x^{3/2}\right)$$

$$= \frac{4}{3} \cdot \frac{3}{2}x^{3/2-1}$$

$$= \frac{4}{3} \cdot \frac{3}{2}x^{1/2}$$

$$= 2x^{1/2} = 2\sqrt{x}$$

**Step 3: Calculate $\left(\frac{dy}{dx}\right)^2$:**

$$\left(f'(x)\right)^2 = (2\sqrt{x})^2 = 4x$$

**Step 4: Set up the arc length integral:**

$$L = \int_0^1 \sqrt{1 + 4x} \, dx$$

**Step 5: Solve the integral using substitution:**

Let $u = 1 + 4x$

Then $du = 4dx$, so $dx = \frac{du}{4}$

When $x = 0$: $u = 1 + 4(0) = 1$

When $x = 1$: $u = 1 + 4(1) = 5$

**Step 6: Substitute and integrate:**

$$L = \int_1^5 \sqrt{u} \cdot \frac{du}{4}$$

$$= \frac{1}{4} \int_1^5 u^{1/2} \, du$$

$$= \frac{1}{4} \left[\frac{u^{1/2+1}}{1/2+1}\right]_1^5$$

$$= \frac{1}{4} \left[\frac{u^{3/2}}{3/2}\right]_1^5$$

$$= \frac{1}{4} \cdot \frac{2}{3} \left[u^{3/2}\right]_1^5$$

$$= \frac{1}{6} \left[u^{3/2}\right]_1^5$$

**Step 7: Evaluate the definite integral:**

$$L = \frac{1}{6} \left[u^{3/2}\right]_1^5$$

$$= \frac{1}{6} \left(5^{3/2} - 1^{3/2}\right)$$

$$= \frac{1}{6} \left(5\sqrt{5} - 1\right)$$

$$= \frac{1}{6} \left(5\sqrt{5} - 1\right)$$

**Step 8: Simplify the result:**

$$5^{3/2} = 5 \cdot 5^{1/2} = 5\sqrt{5}$$

Since $\sqrt{5} = \sqrt{5}$, we can also write this as:

$$L = \frac{1}{6}(5\sqrt{5} - 1)$$

**Alternative form:**

We can rationalize or express this in different ways:
$$L = \frac{5\sqrt{5} - 1}{6}$$

**Numerical approximation:**

$$L = \frac{5\sqrt{5} - 1}{6} \approx \frac{5(2.236) - 1}{6} \approx \frac{11.180 - 1}{6} \approx \frac{10.180}{6} \approx 1.697$$

**Q1) Calculate the arc length of $y = \ln(\sec x)$ between $0 \leq x \leq \pi/4$.**

**Answer:** $\ln(\sqrt{2} + 1)$ units <br>
*Explanation: We use the arc length formula for a function y = f(x).*

$$\text{Given: } y = \ln(\sec x) \text{ over the interval } [0, \pi/4]$$

**Step 1: Find the derivative of y with respect to x:**

$$y = \ln(\sec x)$$

$$\frac{dy}{dx} = \frac{d}{dx}[\ln(\sec x)]$$

Using the chain rule:
$$\frac{dy}{dx} = \frac{1}{\sec x} \cdot \frac{d}{dx}(\sec x)$$

$$= \frac{1}{\sec x} \cdot \sec x \tan x$$

$$= \tan x$$

**Step 2: Apply the arc length formula:**

$$L = \int_0^{\pi/4} \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

$$= \int_0^{\pi/4} \sqrt{1 + \tan^2 x} \, dx$$

**Step 3: Use the trigonometric identity:**

$$1 + \tan^2 x = \sec^2 x$$

$$L = \int_0^{\pi/4} \sqrt{\sec^2 x} \, dx$$

$$= \int_0^{\pi/4} |\sec x| \, dx$$

Since $\sec x > 0$ for $x \in [0, \pi/4]$:

$$= \int_0^{\pi/4} \sec x \, dx$$

**Step 4: Integrate secant:**

$$\int \sec x \, dx = \ln|\sec x + \tan x| + C$$

**Step 5: Evaluate the definite integral:**

$$L = \left[\ln|\sec x + \tan x|\right]_0^{\pi/4}$$

At $x = \pi/4$:
- $\sec(\pi/4) = \sqrt{2}$
- $\tan(\pi/4) = 1$

At $x = 0$:
- $\sec(0) = 1$
- $\tan(0) = 0$

$$L = \ln|\sqrt{2} + 1| - \ln|1 + 0|$$

$$= \ln(\sqrt{2} + 1) - \ln(1)$$

$$= \ln(\sqrt{2} + 1) - 0$$

$$= \ln(\sqrt{2} + 1)$$

**Method 4: Revolution around horizontal line y = k**

### **Integration: Properties of Gradients**

The gradient is another word for the derivative, but it's defined as a **vector field**. The gradient function gives the slope of a function at any single point on its curve, but extends this concept to multivariable functions.

The gradient is found by applying the vector operator **** (nabla) to a scalar function. This creates what's called a **gradient vector field**.

### **Gradient Formulas**

#### **Two Dimensions**
For a function $f(x,y)$:
$$\text{grad } f(x,y) = \nabla f(x,y) = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j}$$

#### **Three Dimensions**
For a function $f(x,y,z)$:
$$\text{grad } f(x,y,z) = \nabla f(x,y,z) = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j} + \frac{\partial f}{\partial z}\mathbf{k}$$

### **Properties of Gradients**

1. **Input/Output:** Takes a scalar function and produces a vector function
2. **Dimensionality:** The gradient vector $\nabla f(x,y)$ lies in the same dimensional space as the input
3. **Direction:** Points in the direction of steepest increase of the function
4. **Magnitude:** Represents the rate of change in that direction

#### **Example 1: Two-Dimensional Gradient**

**Find the gradient of $f(x,y) = x + 3y^2$**

**Step 1:** Apply the gradient formula
$$\nabla f(x,y) = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j}$$

**Step 2:** Find partial derivatives
$$\frac{\partial}{\partial x}(x + 3y^2) = 1 + 0 = 1$$
$$\frac{\partial}{\partial y}(x + 3y^2) = 0 + 6y = 6y$$

**Step 3:** Combine results
$$\nabla f(x,y) = 1\mathbf{i} + 6y\mathbf{j} = \mathbf{i} + 6y\mathbf{j}$$

**Answer:** $\nabla f(x,y) = \mathbf{i} + 6y\mathbf{j}$

#### **Example 2: Three-Dimensional Gradient**

**Find the gradient of $f(x,y,z) = \sin(x) \cdot e^y \cdot \ln(z)$**

**Solution:**

**Step 1:** Apply the 3D gradient formula
$$\nabla f(x,y,z) = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j} + \frac{\partial f}{\partial z}\mathbf{k}$$

**Step 2:** Find each partial derivative

For $\frac{\partial f}{\partial x}$:
$$\frac{\partial f}{\partial x} = \cos(x) \cdot e^y \cdot \ln(z)$$

For $\frac{\partial f}{\partial y}$:
$$\frac{\partial f}{\partial y} = \sin(x) \cdot e^y \cdot \ln(z)$$

For $\frac{\partial f}{\partial z}$:
$$\frac{\partial f}{\partial z} = \sin(x) \cdot e^y \cdot \frac{1}{z}$$

**Step 3:** Combine results
$$\nabla f(x,y,z) = \cos(x)e^y\ln(z)\mathbf{i} + \sin(x)e^y\ln(z)\mathbf{j} + \frac{\sin(x)e^y}{z}\mathbf{k}$$

### **Gradient Algorithms and Local/Global Extrema**

### **Gradient Descent Algorithm**

Gradient Descent is a popular optimization algorithm in Machine Learning used for **minimizing cost or loss functions**. It's a first-order iterative algorithm that numerically finds where a function outputs its lowest values.

#### **Key Features**
- Finds **local minima** and approximates solutions numerically
- **No formula required** - just needs the numerical output of the function
- Works by taking steps in the **opposite direction** of the gradient (steepest descent)

#### **How It Works**

The algorithm repeatedly takes steps in the direction **opposite** to the function's gradient at the current point, moving toward the steepest descent.

**Note:** If we step in the direction **of** the gradient, it leads to a local maximum - this is called **gradient ascent**.


#### **Step Size () - The Critical Factor**

The step size controls whether the algorithm:
- **Too large:** Might not converge or take too many steps
- **Too small:** Takes too long to converge
- **Just right:** Converges quickly to the minimum

#### **Simple Example**

For the function $f(x) = x^2 - 6x$:
- **Analytical solution:** Set $f'(x) = 0 \Rightarrow 2x - 6 = 0 \Rightarrow x = 3$
- **Gradient descent solution:** $x  2.99999967$ (very close to 3!)

**Why use gradient descent?** For complex functions, setting $f'(x) = 0$ isn't always easy to solve!

### **Local and Global Maxima and Minima**



#### **Definitions**

**Local Maximum:** A point $(x,y)$ where the $y$-coordinate is larger than all nearby points.

**Local Minimum:** A point $(x,y)$ where the $y$-coordinate is smaller than all nearby points.

**Local Extremum:** Either a local minimum or maximum.

#### **Finding Extrema: The Derivative Test**

**Key Principle:** If $(x,y)$ is an extremum point and the derivative exists, then the tangent line is horizontal.

**Mathematical Rule:** Extrema occur where $f'(x) = 0$

**Important Note:** Not all points where $f'(x) = 0$ are extrema - some are **saddle points**!

#### **Types of Critical Points**



| **Point Type** | **Characteristic** | **Nearby Points** |
|----------------|-------------------|-------------------|
| **Local Maximum** | Highest point locally | All nearby points have smaller $y$-values |
| **Local Minimum** | Lowest point locally | All nearby points have larger $y$-values |
| **Saddle Point** | Horizontal tangent but not extremum | Some nearby points higher, some lower |

#### **How to Identify Extrema**

**Step 1:** Find where $f'(x) = 0$

**Step 2:** Test points on both sides of the critical point:
- If both sides have **smaller** $y$-values  **Local Maximum**
- If both sides have **larger** $y$-values  **Local Minimum**  
- If sides have **mixed** $y$-values  **Saddle Point**

### **Worked Examples**

#### **Example 1: $y = x^2 - x$**

**Step 1:** Find the derivative
$$y' = 2x - 1$$

**Step 2:** Set $y' = 0$
$$2x - 1 = 0 \Rightarrow x = \frac{1}{2}$$

**Step 3:** Find the $y$-coordinate
$$y = \left(\frac{1}{2}\right)^2 - \frac{1}{2} = \frac{1}{4} - \frac{1}{2} = -\frac{1}{4}$$

**Step 4:** Test nearby points
- At $x = 0$: $y = 0$
- At $x = 1$: $y = 0$

Both values are greater than $-\frac{1}{4}$.

**Answer:** Local minimum at $\left(\frac{1}{2}, -\frac{1}{4}\right)$

#### **Example 2: $y = 2 + 3x - x^3$**

**Step 1:** Find the derivative
$$y' = 3 - 3x^2$$

**Step 2:** Set $y' = 0$
$$3 - 3x^2 = 0 \Rightarrow x^2 = 1 \Rightarrow x = 1 \text{ or } x = -1$$

**Step 3:** Analyze $x = 1$
- At $x = 1$: $y = 2 + 3(1) - 1^3 = 4$
- Test points: $x = 0 \Rightarrow y = 2$, $x = 2 \Rightarrow y = 0$
- Both nearby values are smaller  **Local maximum at $(1, 4)$**

**Step 4:** Analyze $x = -1$
- At $x = -1$: $y = 2 + 3(-1) - (-1)^3 = 0$  
- Test points: $x = -2 \Rightarrow y = 4$, $x = 0 \Rightarrow y = 2$
- Both nearby values are larger  **Local minimum at $(-1, 0)$**

### **Second Derivative Test**

The **Second Derivative Test** is another methodology to determine whether a critical point is a local maximum or minimum. Instead of testing nearby points, we examine how the slope itself is changing.

#### **Understanding the Concept**

**First Derivative:** Finds the slope of a function
**Second Derivative:** Finds the slope of the slope (how the slope is changing)

#### **Mathematical Relationship**
- **First derivative** $f'(x)$ = rate of change of the function
- **Second derivative** $f''(x)$ = rate of change of the slope

#### **The Second Derivative Test Rules**

For a critical point where $f'(x) = 0$:

| **Second Derivative** | **Result** | **Meaning** |
|----------------------|------------|-------------|
| $f''(x) < 0$ | **Local Maximum** | Slope is decreasing through the critical point |
| $f''(x) > 0$ | **Local Minimum** | Slope is increasing through the critical point |
| $f''(x) = 0$ | **Test Fails** | Need alternative methods (could be inflection point) |

#### **Case 1: Local Minimum ($f''(x) > 0$)**
When the second derivative is **positive**:
- **Left side:** Slope is negative (decreasing function)
- **Critical point:** Slope is zero (flat/horizontal)
- **Right side:** Slope is positive (increasing function)
- **Pattern:** Slope goes from negative  zero  positive (increasing slope)

#### **Case 2: Local Maximum ($f''(x) < 0$)**
When the second derivative is **negative**:
- **Left side:** Slope is positive (increasing function)
- **Critical point:** Slope is zero (flat/horizontal)  
- **Right side:** Slope is negative (decreasing function)
- **Pattern:** Slope goes from positive  zero  negative (decreasing slope)

### **Worked Example**

**Function:** $y = x^2 - x$

**Step 1:** Find the first derivative
$$y' = 2x - 1$$

**Step 2:** Find critical points (set $y' = 0$)
$$2x - 1 = 0 \Rightarrow x = \frac{1}{2}$$

**Step 3:** Find the second derivative
$$y'' = 2$$

**Step 4:** Apply the second derivative test
At $x = \frac{1}{2}$: $y'' = 2 > 0$

**Conclusion:** Since $y'' > 0$, the critical point at $x = \frac{1}{2}$ is a **local minimum**.

**Step 5:** Find the coordinates
$$y = \left(\frac{1}{2}\right)^2 - \frac{1}{2} = -\frac{1}{4}$$

**Answer:** Local minimum at $\left(\frac{1}{2}, -\frac{1}{4}\right)$



### **Slope Behavior Analysis**

### For the Example $y = x^2 - x$

Since $y'' = 2$ (positive constant), the slope is **continually increasing**:

- **Moving left to right:** 
  - Slope starts negative (function decreasing)
  - Slope reaches zero (horizontal tangent)
  - Slope becomes positive (function increasing)

This confirms our conclusion: the critical point is a **minimum** because the slope increases through it.

### **Integration Optimization**

Optimization involves finding the **largest or smallest value** that a function can take under various **constraints**. In real-world scenarios, businesses often need to maximize profit or minimize cost while satisfying certain limitations.

#### **Constraints**
- Limited raw materials
- Insufficient workforce
- Budget restrictions
- Time limitations
- Physical space constraints

### **Example 1: Maximizing Building Area**

#### **Problem Statement**
A construction company must build a community hall with these constraints:
- **Budget limitation:** Standard specifications only
- **Wall constraint:** Total wall length cannot exceed 300 meters
- **Design:** Rectangular hall with one side fully open for glass windows
- **Goal:** Maximize the area of the hall

**Step 1: Set up variables**
- Let $x$ = length of the two closed sides
- Let $y$ = length of the one closed side (parallel to the open side)
- Both $x > 0$ and $y > 0$

**Step 2: Establish the constraint**
Since only three walls are needed:
$$2x + y = 300$$
Therefore: $y = 300 - 2x$

**Step 3: Create the objective function**
Area to maximize:
$$A = x \cdot y = x(300 - 2x) = 300x - 2x^2$$

**Step 4: Determine the domain**
For $y > 0$: $300 - 2x > 0 \Rightarrow x < 150$
Domain: $0 < x < 150$

**Step 5: Find the maximum**
$$A' = 300 - 4x$$
Set $A' = 0$: $300 - 4x = 0 \Rightarrow x = 75$

**Step 6: Calculate dimensions and area**
- $x = 75$ meters
- $y = 300 - 2(75) = 150$ meters  
- **Maximum area:** $A = 75 \times 150 = 11,250$ m

**Optimal dimensions:** 75m  150m for a maximum area of 11,250 m

### **Example 2: Maximizing Car Rental Revenue**

#### **Problem Statement**
A car rental company needs to determine the optimal daily rental rate:
- **Rate range:** $50  x  200$ (dollars per day)
- **Demand function:** $n(x) = 1000 - 5x$ (number of cars rented per day)
- **Goal:** Maximize daily revenue

**Step 1: Set up the revenue function**
$$R(x) = \text{(rate per car)} \times \text{(number of cars)}$$
$$R(x) = x \cdot n(x) = x(1000 - 5x)$$
$$R(x) = 1000x - 5x^2$$

**Step 2: Find the critical point**
$$R'(x) = 1000 - 10x$$
Set $R'(x) = 0$: $1000 - 10x = 0 \Rightarrow x = 100$

**Step 3: Check if critical point is in the domain**
Since $50  x  200$ and $x = 100$, the critical point is valid.

**Step 4: Evaluate revenue at critical point and endpoints**
- At $x = 100$: $R(100) = 1000(100) - 5(100)^2 = 50,000$
- At $x = 50$: $R(50) = 1000(50) - 5(50)^2 = 37,500$
- At $x = 200$: $R(200) = 1000(200) - 5(200)^2 = 0$

**Step 5: Compare values**
$R(100) = 50,000 > R(50) = 37,500 > R(200) = 0$

**Optimal daily rental rate:** $100 for maximum revenue of $50,000

<br><br>

### **Optimization Strategy Summary**

1. **Define variables** for the quantities you want to optimize
2. **Identify constraints** and express them as equations
3. **Create objective function** (what you want to maximize/minimize)
4. **Use constraints** to eliminate variables and create a single-variable function
5. **Find critical points** using derivatives
6. **Check endpoints** if the domain is bounded
7. **Compare values** to find the absolute maximum/minimum

### **Mathematical Tools Used**
- **Extreme Value Theorem:** Guarantees maximum/minimum exists in closed intervals
- **First Derivative Test:** Finds critical points where $f'(x) = 0$
- **Second Derivative Test:** Confirms whether critical points are maxima or minima
- **Endpoint Analysis:** Compares values at domain boundaries

$$y = \frac{x^2 - 1}{x}$$

$$\frac{1}{x^2} = -1$$

**Analysis:** Since $\frac{1}{x^2} > 0$ for all $x \neq 0$, there are **no critical points** where $y' = 0$.

**Step 4:** Analyze the behavior
- $y' = 1 + \frac{1}{x^2} > 0$ for all $x \neq 0$
- The function is always increasing where it's defined
- Note: $x = 0$ is not in the domain (vertical asymptote)

**Answer:** **No local extrema** exist for this function.

### **Problem 2**
$$y = \cos(2x) - x$$

**Step 1:** Find the first derivative
$$y' = \frac{d}{dx}[\cos(2x) - x] = -\sin(2x) \cdot 2 - 1 = -2\sin(2x) - 1$$

**Step 2:** Find critical points (set $y' = 0$)
$$-2\sin(2x) - 1 = 0$$
$$\sin(2x) = -\frac{1}{2}$$

**Step 3:** Solve for $x$
$$2x = \arcsin\left(-\frac{1}{2}\right) + 2\pi k \text{ or } 2x = \pi - \arcsin\left(-\frac{1}{2}\right) + 2\pi k$$

Since $\arcsin\left(-\frac{1}{2}\right) = -\frac{\pi}{6}$:
$$2x = -\frac{\pi}{6} + 2\pi k \text{ or } 2x = \pi + \frac{\pi}{6} + 2\pi k$$
$$x = -\frac{\pi}{12} + \pi k \text{ or } x = \frac{7\pi}{12} + \pi k$$

**Step 4:** Apply second derivative test
$$y'' = \frac{d}{dx}[-2\sin(2x) - 1] = -2\cos(2x) \cdot 2 = -4\cos(2x)$$

**For $x = -\frac{\pi}{12} + \pi k$:**
$$y''\left(-\frac{\pi}{12}\right) = -4\cos\left(-\frac{\pi}{6}\right) = -4 \cdot \frac{\sqrt{3}}{2} = -2\sqrt{3} < 0$$
**Local maximum**

**For $x = \frac{7\pi}{12} + \pi k$:**
$$y''\left(\frac{7\pi}{12}\right) = -4\cos\left(\frac{7\pi}{6}\right) = -4 \cdot \left(-\frac{\sqrt{3}}{2}\right) = 2\sqrt{3} > 0$$
**Local minimum**

**Answer:** 
- **Local maxima** at $x = -\frac{\pi}{12} + \pi k$
- **Local minima** at $x = \frac{7\pi}{12} + \pi k$

#### **Problem 2**
$$y = x^3 - 6x^2 + 12x - 5$$

### Solution:

**Step 1:** Find the first derivative
$$y' = \frac{d}{dx}[x^3 - 6x^2 + 12x - 5] = 3x^2 - 12x + 12$$

### **Worksheet Example: Integration Optimization**

#### **Problem 1: Revenue Maximization**

**Given:** A demand equation $x = \sqrt{300 - 0.5y}$, where $x$ is unit price and $y$ is the number of units manufactured per week.

**Question:** How many units should be manufactured and sold each week to maximize the total revenue $(R)$?

### Solution:

**Step 1:** Solve for $y$ in terms of $x$

Starting with: $x = \sqrt{300 - 0.5y}$

Square both sides: $x^2 = 300 - 0.5y$

Rearrange: $0.5y = 300 - x^2$

Therefore: $y = 600 - 2x^2$

**Step 2:** Set up the revenue function

Revenue = number of units  unit price
$$R = y \times x = (600 - 2x^2) \times x = 600x - 2x^3$$

**Step 3:** Find the critical points

$$R' = \frac{d}{dx}[600x - 2x^3] = 600 - 6x^2$$

Set $R' = 0$:
$$600 - 6x^2 = 0$$
$$6x^2 = 600$$
$$x^2 = 100$$
$$x = 10$$ 
(taking positive value since price must be positive)

**Step 4:** Verify this is a maximum using second derivative test

$$R'' = \frac{d}{dx}[600 - 6x^2] = -12x$$

At $x = 10$: $R''(10) = -12(10) = -120 < 0$

Since $R'' < 0$, this confirms a maximum.

**Step 5:** Find the number of units

$$y = 600 - 2x^2 = 600 - 2(10)^2 = 600 - 200 = 400$$

**Answer:** **400 units** should be manufactured and sold each week to maximize revenue.

#### **Problem 2: Investment Account Optimization**

**Given:** The amount of money in an account is given by:
$$y(t) = -0.25t^4 - \frac{5}{3}t^3 - 18t^2 + 180t + 1200$$

where $t > 0$ is the number of years, and $y$ is measured in thousands of dollars.

**Question:** Determine the year during which the value of the account is maximal.

**Step 1:** Find the first derivative

$$\frac{dy}{dt} = \frac{d}{dt}\left[-0.25t^4 - \frac{5}{3}t^3 - 18t^2 + 180t + 1200\right]$$
$$= -t^3 - 5t^2 - 36t + 180$$

**Step 2:** Find critical points (set $\frac{dy}{dt} = 0$)

$$-t^3 - 5t^2 - 36t + 180 = 0$$
$$t^3 + 5t^2 + 36t - 180 = 0$$

**Step 3:** Factor the cubic equation

By testing values or factoring:
$$t^3 + 5t^2 + 36t - 180 = (t - 5)(t^2 + 10t + 36) = 0$$

From $(t - 5) = 0$: $t = 5$

From $t^2 + 10t + 36 = 0$: Using the quadratic formula:
$$t = \frac{-10 \pm \sqrt{100 - 144}}{2} = \frac{-10 \pm \sqrt{-44}}{2}$$

Since the discriminant is negative, these are complex roots.

Therefore, the only real critical point is $t = 5$.

**Step 4:** Apply second derivative test

$$\frac{d^2y}{dt^2} = \frac{d}{dt}[-t^3 - 5t^2 - 36t + 180] = -3t^2 - 10t - 36$$

At $t = 5$:
$$\frac{d^2y}{dt^2}\bigg|_{t=5} = -3(25) - 10(5) - 36 = -75 - 50 - 36 = -161 < 0$$

Since the second derivative is negative, $t = 5$ gives a maximum.

**Step 5:** Verify by checking nearby values

At $t = 5$:
$$y(5) = -0.25(625) - \frac{5}{3}(125) - 18(25) + 180(5) + 1200$$
$$= -156.25 - 208.33 - 450 + 900 + 1200 \approx 1285.42$$

At $t = 6$:
$$y(6) = -0.25(1296) - \frac{5}{3}(216) - 18(36) + 180(6) + 1200$$
$$= -324 - 360 - 648 + 1080 + 1200 = 948$$

Since $y(5) > y(6)$, this confirms $t = 5$ is indeed a maximum.

**Answer:** The value of the account is maximal at **5 years**.
