### **Integration: Finding an Area Under a Curve**

Integrals calculate the area under a curve by breaking it into thin rectangles and summing their areas.



### The Rectangle Method
- Split the area under a curve into many thin vertical rectangles
- Each rectangle has:
  - **Width:** Small change in $x$
  - **Height:** Function value $f(x)$
- Sum all rectangle areas to get the total area

### Mathematical Formula
For area between curve $f(x)$ and x-axis from $x = a$ to $x = b$:
$$\text{Area} = \int_a^b f(x) \, dx$$

### Example: $f(x) = x^2$ from $x = -1$ to $x = 2$

**Solution:**
$$\text{Area} = \int_{-1}^{2} x^2 \, dx = \left[\frac{x^3}{3}\right]_{-1}^{2}$$

$$= \frac{2^3}{3} - \frac{(-1)^3}{3} = \frac{8}{3} - \frac{-1}{3} = \frac{8}{3} + \frac{1}{3} = 3$$

**Answer:** 3 square units

### **Integration:Finding Arc Length of a Curve**

Arc length is the distance along a curve between two points. Unlike straight-line distance, it follows the actual path of the curve.

To find the length of a smooth curve, we:
1. Divide the curve into infinitesimally small sections
2. Calculate the length of each small section using Pythagoras' theorem
3. Add up all the small lengths using integration



### **Pythagorean Approach**

For each tiny triangle along the curve:
- **Horizontal leg:** $\Delta x_i$
- **Vertical leg:** $\Delta y_i$ 
- **Hypotenuse:** $\sqrt{\Delta x_i^2 + \Delta y_i^2}$

### **Arc Length Formula Development**

**Step 1:** Sum of all small segments
$$\text{Arc Length} = \int_a^b \sqrt{\Delta x_i^2 + \Delta y_i^2}$$

**Step 2:** Convert to differential form
$$\text{Arc Length} = \int_a^b \sqrt{dx^2 + dy^2}$$

**Step 3:** Factor out $dx$
$$\text{Arc Length} = \int_a^b \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

**Step 4:** Final formula (since $f'(x) = \frac{dy}{dx}$)
$$s = \int_a^b \sqrt{1 + [f'(x)]^2} \, dx$$

### **Worked Example**

**Find the arc length of $f(x) = \frac{x^2}{2} - \frac{\ln(x)}{4}$ from $x = 2$ to $x = 4$**

**Step 1:** Find the derivative
$$f'(x) = \frac{d}{dx}\left(\frac{x^2}{2} - \frac{\ln(x)}{4}\right) = x - \frac{1}{4x}$$

**Step 2:** Calculate $[f'(x)]^2$
$$[f'(x)]^2 = \left(x - \frac{1}{4x}\right)^2 = x^2 - \frac{1}{2} + \frac{1}{16x^2}$$

**Step 3:** Simplify $1 + [f'(x)]^2$
$$1 + [f'(x)]^2 = 1 + x^2 - \frac{1}{2} + \frac{1}{16x^2} = x^2 + \frac{1}{2} + \frac{1}{16x^2}$$

**Step 4:** Factor the expression
$$1 + [f'(x)]^2 = \left(x + \frac{1}{4x}\right)^2$$

**Step 5:** Set up the integral
$$s = \int_2^4 \sqrt{\left(x + \frac{1}{4x}\right)^2} \, dx = \int_2^4 \left(x + \frac{1}{4x}\right) dx$$

**Step 6:** Integrate
$$s = \int_2^4 x \, dx + \int_2^4 \frac{1}{4x} \, dx = \left[\frac{x^2}{2} + \frac{\ln x}{4}\right]_2^4$$

**Step 7:** Evaluate
$$s = \left(\frac{16}{2} + \frac{\ln 4}{4}\right) - \left(\frac{4}{2} + \frac{\ln 2}{4}\right)$$
$$s = 8 + \frac{\ln 4}{4} - 2 - \frac{\ln 2}{4} = 6 + \frac{\ln 4 - \ln 2}{4}$$
$$s = 6 + \frac{\ln 2}{4}$$

**Answer:** The arc length is $6 + \frac{\ln 2}{4}$ units.

### **Integration: Finding the Area of a Surface**

Surface area is the total area of the outer layer of any object. For curved surfaces, calculating this area is more complex than for standard shapes like cylinders, spheres, or cubes. Finding surface area extends the concept of arc length from one dimension (curves) to two dimensions (surfaces).

To find the area of a curved surface:

1. **Divide** the surface into small subregions
2. **Map** each subregion onto a tangent plane
3. **Calculate** the area of each small piece
4. **Sum** all the small areas using double integrals

#### **Understanding the Process**

For each small subregion (the $i$-th piece):
- **Dimensions:** $\Delta x_i$ and $\Delta y_i$
- **Area:** $\Delta A_i = \Delta x_i \cdot \Delta y_i$

Since we're dealing with a surface, we need to consider approximations in **both** the $x$ and $y$ directions, which is why we use double integrals ($\iint$).

#### **Surface Area Formula**

For a surface defined by $z = f(x,y)$, where the partial derivatives $f_x$ and $f_y$ are continuous over a closed, bounded region $R$:

$$S = \iint_R dS = \iint_R \sqrt{1 + f_x(x,y)^2 + f_y(x,y)^2} \, dA$$

#### **Formula Components**

- **$S$** = Total surface area
- **$f_x(x,y)$** = Partial derivative with respect to $x$
- **$f_y(x,y)$** = Partial derivative with respect to $y$
- **$R$** = The region in the $xy$-plane
- **$dA$** = Small area element in the $xy$-plane

#### **Connection to Arc Length**

| **Arc Length (1D)** | **Surface Area (2D)** |
|---------------------|------------------------|
| $s = \int_a^b \sqrt{1 + [f'(x)]^2} \, dx$ | $S = \iint_R \sqrt{1 + f_x^2 + f_y^2} \, dA$ |
| Uses single integral | Uses double integral |
| One derivative $f'(x)$ | Two partial derivatives $f_x, f_y$ |

### **Worksheet Question**

**Q1:** Find the area under the curve $f(x) = x(x-2)(x+2)$ for $-1 \leq x \leq 1$.

*Explanation: To find the area under the curve, we need to evaluate the definite integral. Since the function is negative in the given interval, we'll need to take the absolute value to get the actual area.*

$$\text{Given: } f(x) = x(x-2)(x+2) \text{ for } -1 \leq x \leq 1$$

**Step 1: Expand the function:**

$$f(x) = x(x-2)(x+2)$$

$$= x[(x-2)(x+2)]$$

$$= x[x^2 - 4]$$

$$= x^3 - 4x$$

**Step 2: Analyze the function's behavior in the interval $[-1, 1]$:**

$$\text{At } x = -1: f(-1) = (-1)^3 - 4(-1) = -1 + 4 = 3$$

$$\text{At } x = 0: f(0) = 0^3 - 4(0) = 0$$

$$\text{At } x = 1: f(1) = 1^3 - 4(1) = 1 - 4 = -3$$

**Step 3: Check if the function changes sign in the interval:**

From the graph and our calculations, we can see that:
- $f(x) > 0$ for $x \in [-1, 0)$
- $f(x) < 0$ for $x \in (0, 1]$

**Step 4: Set up the integral for the area:**

Since the function changes sign at $x = 0$, we need to split the integral:

$$\text{Area} = \int_{-1}^{0} |f(x)| dx + \int_{0}^{1} |f(x)| dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx + \int_{0}^{1} |x^3 - 4x| dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx + \int_{0}^{1} -(x^3 - 4x) dx$$

$$= \int_{-1}^{0} (x^3 - 4x) dx - \int_{0}^{1} (x^3 - 4x) dx$$

**Step 5: Find the antiderivative:**

$$\int (x^3 - 4x) dx = \frac{x^4}{4} - 4 \cdot \frac{x^2}{2} = \frac{x^4}{4} - 2x^2$$

**Step 6: Evaluate the first integral:**

$$\int_{-1}^{0} (x^3 - 4x) dx = \left[\frac{x^4}{4} - 2x^2\right]_{-1}^{0}$$

$$= \left(\frac{0^4}{4} - 2(0)^2\right) - \left(\frac{(-1)^4}{4} - 2(-1)^2\right)$$

$$= 0 - \left(\frac{1}{4} - 2\right)$$

$$= 0 - \left(\frac{1}{4} - \frac{8}{4}\right)$$

$$= 0 - \left(-\frac{7}{4}\right) = \frac{7}{4}$$

**Step 7: Evaluate the second integral:**

$$\int_{0}^{1} (x^3 - 4x) dx = \left[\frac{x^4}{4} - 2x^2\right]_{0}^{1}$$

$$= \left(\frac{1^4}{4} - 2(1)^2\right) - \left(\frac{0^4}{4} - 2(0)^2\right)$$

$$= \left(\frac{1}{4} - 2\right) - 0$$

$$= \frac{1}{4} - \frac{8}{4} = -\frac{7}{4}$$

**Step 8: Calculate the total area:**

$$\text{Area} = \int_{-1}^{0} (x^3 - 4x) dx - \int_{0}^{1} (x^3 - 4x) dx$$

$$= \frac{7}{4} - \left(-\frac{7}{4}\right)$$

$$= \frac{7}{4} + \frac{7}{4}$$

$$= \frac{14}{4} = \frac{7}{2}$$

**Final Answer:** The area under the curve is $\frac{7}{2}$ square units.

**Q2:** Calculate the arc length of the graph of $f(x) = \frac{4}{3}x^{3/2}$ over the interval $[0, 1]$.

*Explanation: To find the arc length of a curve, we use the arc length formula for a function y = f(x).*

$$\text{Given: } f(x) = \frac{4}{3}x^{3/2} \text{ over the interval } [0, 1]$$

**Step 1: Recall the arc length formula:**

For a function $y = f(x)$ over the interval $[a, b]$, the arc length is:

$$L = \int_a^b \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

**Step 2: Find the derivative of f(x):**

$$f(x) = \frac{4}{3}x^{3/2}$$

$$f'(x) = \frac{d}{dx}\left(\frac{4}{3}x^{3/2}\right)$$

$$= \frac{4}{3} \cdot \frac{3}{2}x^{3/2-1}$$

$$= \frac{4}{3} \cdot \frac{3}{2}x^{1/2}$$

$$= 2x^{1/2} = 2\sqrt{x}$$

**Step 3: Calculate $\left(\frac{dy}{dx}\right)^2$:**

$$\left(f'(x)\right)^2 = (2\sqrt{x})^2 = 4x$$

**Step 4: Set up the arc length integral:**

$$L = \int_0^1 \sqrt{1 + 4x} \, dx$$

**Step 5: Solve the integral using substitution:**

Let $u = 1 + 4x$

Then $du = 4dx$, so $dx = \frac{du}{4}$

When $x = 0$: $u = 1 + 4(0) = 1$

When $x = 1$: $u = 1 + 4(1) = 5$

**Step 6: Substitute and integrate:**

$$L = \int_1^5 \sqrt{u} \cdot \frac{du}{4}$$

$$= \frac{1}{4} \int_1^5 u^{1/2} \, du$$

$$= \frac{1}{4} \left[\frac{u^{1/2+1}}{1/2+1}\right]_1^5$$

$$= \frac{1}{4} \left[\frac{u^{3/2}}{3/2}\right]_1^5$$

$$= \frac{1}{4} \cdot \frac{2}{3} \left[u^{3/2}\right]_1^5$$

$$= \frac{1}{6} \left[u^{3/2}\right]_1^5$$

**Step 7: Evaluate the definite integral:**

$$L = \frac{1}{6} \left[u^{3/2}\right]_1^5$$

$$= \frac{1}{6} \left(5^{3/2} - 1^{3/2}\right)$$

$$= \frac{1}{6} \left(5\sqrt{5} - 1\right)$$

$$= \frac{1}{6} \left(5\sqrt{5} - 1\right)$$

**Step 8: Simplify the result:**

$$5^{3/2} = 5 \cdot 5^{1/2} = 5\sqrt{5}$$

Since $\sqrt{5} = \sqrt{5}$, we can also write this as:

$$L = \frac{1}{6}(5\sqrt{5} - 1)$$

**Alternative form:**

We can rationalize or express this in different ways:
$$L = \frac{5\sqrt{5} - 1}{6}$$

**Numerical approximation:**

$$L = \frac{5\sqrt{5} - 1}{6} \approx \frac{5(2.236) - 1}{6} \approx \frac{11.180 - 1}{6} \approx \frac{10.180}{6} \approx 1.697$$

*Note: There appears to be a discrepancy between the calculated result and the given answer form. The calculation above gives $\frac{5\sqrt{5} - 1}{6}$, which may need verification with the original problem statement or methodology reference.*

**Final Answer:** Based on the standard arc length calculation, the arc length is $\frac{5\sqrt{5} - 1}{6}$ units.

**Q1) Calculate the arc length of $y = \ln(\sec x)$ between $0 \leq x \leq \pi/4$.**

**Answer:** $\ln(\sqrt{2} + 1)$ units <br>
*Explanation: We use the arc length formula for a function y = f(x).*

$$\text{Given: } y = \ln(\sec x) \text{ over the interval } [0, \pi/4]$$

**Step 1: Find the derivative of y with respect to x:**

$$y = \ln(\sec x)$$

$$\frac{dy}{dx} = \frac{d}{dx}[\ln(\sec x)]$$

Using the chain rule:
$$\frac{dy}{dx} = \frac{1}{\sec x} \cdot \frac{d}{dx}(\sec x)$$

$$= \frac{1}{\sec x} \cdot \sec x \tan x$$

$$= \tan x$$

**Step 2: Apply the arc length formula:**

$$L = \int_0^{\pi/4} \sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

$$= \int_0^{\pi/4} \sqrt{1 + \tan^2 x} \, dx$$

**Step 3: Use the trigonometric identity:**

$$1 + \tan^2 x = \sec^2 x$$

$$L = \int_0^{\pi/4} \sqrt{\sec^2 x} \, dx$$

$$= \int_0^{\pi/4} |\sec x| \, dx$$

Since $\sec x > 0$ for $x \in [0, \pi/4]$:

$$= \int_0^{\pi/4} \sec x \, dx$$

**Step 4: Integrate secant:**

$$\int \sec x \, dx = \ln|\sec x + \tan x| + C$$

**Step 5: Evaluate the definite integral:**

$$L = \left[\ln|\sec x + \tan x|\right]_0^{\pi/4}$$

At $x = \pi/4$:
- $\sec(\pi/4) = \sqrt{2}$
- $\tan(\pi/4) = 1$

At $x = 0$:
- $\sec(0) = 1$
- $\tan(0) = 0$

$$L = \ln|\sqrt{2} + 1| - \ln|1 + 0|$$

$$= \ln(\sqrt{2} + 1) - \ln(1)$$

$$= \ln(\sqrt{2} + 1) - 0$$

$$= \ln(\sqrt{2} + 1)$$

**Q2) Calculate the arc length of $x = \frac{2}{3}(y - 1)^{3/2}$ between $1 \leq y \leq 4$.**

**Answer:** $\frac{14}{3}$ units <br>
*Explanation: Since x is given as a function of y, we use the arc length formula for x = g(y).*

$$\text{Given: } x = \frac{2}{3}(y - 1)^{3/2} \text{ over the interval } [1, 4]$$

**Step 1: Find the derivative of x with respect to y:**

$$x = \frac{2}{3}(y - 1)^{3/2}$$

$$\frac{dx}{dy} = \frac{2}{3} \cdot \frac{3}{2}(y - 1)^{3/2-1}$$

$$= \frac{2}{3} \cdot \frac{3}{2}(y - 1)^{1/2}$$

$$= (y - 1)^{1/2} = \sqrt{y - 1}$$

**Step 2: Apply the arc length formula for x = g(y):**

$$L = \int_1^4 \sqrt{1 + \left(\frac{dx}{dy}\right)^2} \, dy$$

$$= \int_1^4 \sqrt{1 + (\sqrt{y - 1})^2} \, dy$$

$$= \int_1^4 \sqrt{1 + (y - 1)} \, dy$$

$$= \int_1^4 \sqrt{y} \, dy$$

**Step 3: Integrate:**

$$L = \int_1^4 y^{1/2} \, dy$$

$$= \left[\frac{y^{1/2+1}}{1/2+1}\right]_1^4$$

$$= \left[\frac{y^{3/2}}{3/2}\right]_1^4$$

$$= \frac{2}{3}\left[y^{3/2}\right]_1^4$$

**Step 4: Evaluate the definite integral:**

$$L = \frac{2}{3}\left[y^{3/2}\right]_1^4$$

$$= \frac{2}{3}(4^{3/2} - 1^{3/2})$$

$$= \frac{2}{3}(8 - 1)$$

$$= \frac{2}{3} \cdot 7$$

$$= \frac{14}{3}$$

**Q3) Explore further how surface regions can be calculated for different object shapes.**

**Answer:** Surface area calculations using revolution methods <br>
*Explanation: Surface areas can be calculated using surface of revolution formulas when curves are rotated around axes.*

**Surface of Revolution Formulas:**

**Method 1: Revolution around the x-axis**
When $y = f(x)$ is revolved around the x-axis from $x = a$ to $x = b$:

$$S = 2\pi \int_a^b y\sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

$$S = 2\pi \int_a^b f(x)\sqrt{1 + [f'(x)]^2} \, dx$$

**Method 2: Revolution around the y-axis**
When $x = g(y)$ is revolved around the y-axis from $y = c$ to $y = d$:

$$S = 2\pi \int_c^d x\sqrt{1 + \left(\frac{dx}{dy}\right)^2} \, dy$$

$$S = 2\pi \int_c^d g(y)\sqrt{1 + [g'(y)]^2} \, dy$$

**Method 3: Revolution around vertical line x = k**
When $y = f(x)$ is revolved around the line $x = k$:

$$S = 2\pi \int_a^b |x - k|\sqrt{1 + \left(\frac{dy}{dx}\right)^2} \, dx$$

**Method 4: Revolution around horizontal line y = k**
When $x = g(y)$ is revolved around the line $y = k$:

$$S = 2\pi \int_c^d |y - k|\sqrt{1 + \left(\frac{dx}{dy}\right)^2} \, dy$$
