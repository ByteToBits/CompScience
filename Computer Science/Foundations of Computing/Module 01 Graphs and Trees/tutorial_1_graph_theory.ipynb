#### **Question 4: Minimum Spanning Tree and Binary Search Tree**

##### **Minimum Spanning Tree**

For a **Connected Undirected Graph with Weighted Edgess**, a **Minimum Spanning Tree (MST)** is a subset of the edges in the graph that connects all vertices together (without creating cycles) while minimizing the total edge weight. There are two most well-known algorithm, Prim's (Priority Queue) and Kruskal's (Union-Find - Disjoint-Set) Minimum Spanning Tree Algorithm. It is used in a widely used in Networking (Spanning Tree Protocal) and Clustering Algorithms.

(Source Material: William Fest (https://www.youtube.com/watch?v=JZBQLXgSGfs) & Monash University )



##### **Binary Search Tree**

A Binary Search Tree is an extension of the binary tree with an **Ordering Property**. It is defined by a **Comparison** where all values in the Left subtree are less than the Root Node's Value and all values in the Right Subtree are greater. The **Values of the Nodes Must be Comparable**, no duplicate values are allowed in standard implementations, and In-Order Traversal yields sorted elements. It sees implementation in Database Indexing, Priority Queues, and implementation of maps and sets.



#### **Question 5: Deleting A Node in a Balanced Binary Tree**

**Assume you are deleting the coloured vertex in the following tree. In order to maintain this as a balanced binary tree, how would you rearrange the remaining vertices? Show in a drawing.**

Maintaining balance in a binary search tree is essential because imbalance significantly degrades performance and compromises operational efficiency. Without proper rebalancing after deletion operations, the tree structure gradually deteriorates into a disorganized state, requiring more comparison operations during searches. This inefficiency becomes increasingly problematic as the dataset grows, transforming logarithmic time complexity into linear complexity.
