```python
# Test 1: Set Operations
from sympy import FiniteSet

A = FiniteSet(1, 2, 3)
B = FiniteSet(3, 4, 5)

print("A union B:", A.union(B))
print("A intersect B:", A.intersect(B))
print("A - B:", A - B)
```

```python
# Test 2: Solve an Algebraic Equation
from sympy import symbols, Eq, solve

x = symbols('x')
eq = Eq(x**2 - 4, 0)
solutions = solve(eq, x)
print("Solutions to x^2 - 4 = 0:", solutions)
```

```python
from sympy import Matrix, init_printing
from IPython.display import display
import numpy as np

init_printing()

# Original NumPy matrices
A_np = np.array([[1, 2], [3, 4]])
B_np = np.array([[5], [6]])
C_np = A_np @ B_np

# Convert to SymPy for pretty display
A = Matrix(A_np)
B = Matrix(B_np)
C = Matrix(C_np)

print("Matrix A:")
display(A)

print("Matrix B:")
display(B)

print("A * B:")
display(C)
```

```python
# Test 4: Basic Graph with networkx
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1)])

nx.draw(G, with_labels=True)
plt.title("Triangle Graph")
plt.show()
```

```python
# Test 5: Derivative and Plotting
from sympy import diff, sin
import matplotlib.pyplot as plt
import numpy as np

f = sin(x)
df = diff(f, x)
print("Derivative of sin(x):", df)

# Plotting sin(x) and its derivative
x_vals = np.linspace(0, 2*np.pi, 100)
y_vals = np.sin(x_vals)
dy_vals = np.cos(x_vals)

plt.plot(x_vals, y_vals, label="sin(x)")
plt.plot(x_vals, dy_vals, label="cos(x) (derivative)")
plt.legend()
plt.title("sin(x) and its Derivative")
plt.grid(True)
plt.show()
```
