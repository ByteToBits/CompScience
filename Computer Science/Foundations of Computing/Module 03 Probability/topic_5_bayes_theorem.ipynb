### **Probability Theory: Total Probability and Bayes' Theorem**

### **Total Probability Formula**

The following Venn diagram shows the intersection of the probability of two events. Let E and F be events:



We may express $E$ as:
$$E = E \cap F \cup E \cap F^c$$

For a point to be in $E$, it must either be in both $E$ and $F$ or be in $E$ but not in $F$. 
As $E \cap F$ and $E \cap F^c$ are clearly mutually exclusive, we have $P(E)$ as:

$$P(E) = P(E \cap F) + P(E \cap F^c)$$

Using the definition of conditional probability, this can be rewritten as:

$$P(E) = P(E|F)P(F) + P(E|F^c)P(F^c)$$

Since $P(F^c) = 1 - P(F)$, we get:

$$P(E) = P(E|F)P(F) + P(E|F^c)[1 - P(F)]$$

The probability of the event $E$ is a weighted average of:
- The conditional probability of $E$ given that $F$ has occurred
- The conditional probability of $E$ given that $F$ has not occurred

Each conditional probability is weighted by the probability of the event it is conditioned on. This formula is extremely useful as it often enables us to determine the probability of an event by first 'conditioning' on whether or not some second event has occurred. There are many instances where it is difficult to compute the probability of an event directly, but straightforward once we know whether or not some second event has occurred.

### **Bayes' Theorem**

The formula for Bayes' theorem can be written in various ways. The most common version is:

$$P(A|B) = \frac{P(A)P(B|A)}{P(B)}$$

Where:
- $P(A|B)$ is the conditional probability of event $A$ occurring, given that $B$ is true
- $P(B|A)$ is the conditional probability of event $B$ occurring, given that $A$ is true
- $P(A)$ and $P(B)$ are the probabilities of $A$ and $B$ occurring independently of one another

### **Extended Bayes' Theorem**

This can also be expanded for $n$ mutually exclusive events $A_1, A_2, \ldots, A_n$ whose union is the sample space $S$. 

Let $B$ be any given event. Then Bayes' theorem states:

$$P(A_i|B) = \frac{P(A_i)P(B|A_i)}{\sum_{j=1}^{n}P(A_j)P(B|A_j)}; \quad i=1,2,...,n$$

Where:
- $P(A_1), P(A_2), \ldots, P(A_n)$ are called the prior probabilities of $A_1, A_2, \ldots, A_n$
- $P(A_1|B), P(A_2|B), \ldots, P(A_n|B)$ are called the posterior probabilities of $A_1, A_2, \ldots, A_n$

Also note that the sum of all the posterior probabilities must equal 1.

### **Worksheet Examples**

**Q1) You are planning an outdoor event today but it is cloudy. What is the probability that it will rain today given that it is cloudy?**
**50% of all rainy days start off cloudy. Probability that it is cloudy in the morning is 50%. Probability of seeing rian this month is only 10%**

*Explanation: This problem requires the application of Bayes' theorem to find conditional probability.*

$$\text{Bayes' Theorem: } P(A|B) = \frac{P(A) \times P(B|A)}{P(B)}$$

Given information:
- 60% of all rainy days start off cloudy: $P(Cloudy|Rain) = 0.6$
- Probability that it is cloudy in the morning is 50%: $P(Cloudy) = 0.5$
- Probability of seeing rain this month is only 10%: $P(Rain) = 0.1$

Step 1: Identify the events for Bayes' Theorem application:
- Event A: It will rain today
- Event B: It is cloudy

Step 2: Identify the known probabilities:
$$P(Rain) = 0.1$$
$$P(Cloudy) = 0.5$$
$$P(Cloudy|Rain) = 0.6$$

Step 3: Apply Bayes' Theorem to find the probability of rain given cloudiness:
$$P(Rain|Cloudy) = \frac{P(Rain) \times P(Cloudy|Rain)}{P(Cloudy)}$$

Step 4: Substitute the values and calculate:
$$P(Rain|Cloudy) = \frac{0.1 \times 0.6}{0.5}$$
$$P(Rain|Cloudy) = \frac{0.06}{0.5} = 0.12$$

Therefore, the probability that it will rain today given that it is cloudy is 0.12 or 12%.

**Q2) Some groups of students participate in coding competitions but we are particularly interested in 3 groups as they have previously won. There could be many tasks the groups can complete at each competition.**

**Group 1 completed 8 tasks, 6% of their submission won first prize.**
**Group 2 completed 7 tasks, 10% of their submission won first prize.**
**Group 3 completed 5 tasks, 4% of their submission won first prize.**

**What are the chances that Group 1 will win the first prize?**

*Explanation: This problem requires applying Bayes' theorem for multiple events to find the probability that Group 1 wins, given that a first prize is awarded.*

Step 1: Identify what we're looking for and the relevant probabilities.
- We need to find $P(\text{Group 1} | \text{first prize})$ - the probability that Group 1 wins, given that a first prize is awarded.
- The problem gives us $P(\text{first prize} | \text{Group i})$ for each group, but we need to apply Bayes' theorem.

Step 2: Determine the prior probabilities for each group based on the number of tasks completed.
$$P(\text{Group 1}) = \frac{8}{20} = 0.4$$
$$P(\text{Group 2}) = \frac{7}{20} = 0.35$$
$$P(\text{Group 3}) = \frac{5}{20} = 0.25$$

Where the total number of tasks is $8 + 7 + 5 = 20$.

Step 3: List the conditional probabilities given in the problem.
$$P(\text{first prize} | \text{Group 1}) = 0.06$$
$$P(\text{first prize} | \text{Group 2}) = 0.10$$
$$P(\text{first prize} | \text{Group 3}) = 0.04$$

Step 4: Apply Bayes' theorem for multiple events.
$$P(A_i|B) = \frac{P(A_i)P(B|A_i)}{\sum_{i=1}^{n} P(A_i)P(B|A_i)}; i = 1, 2, ..., n$$

For our problem:
$$P(\text{Group 1} | \text{first prize}) = \frac{P(\text{Group 1})P(\text{first prize}|\text{Group 1})}{\sum_{i=1}^{3} P(\text{Group i})P(\text{first prize}|\text{Group i})}$$

Step 5: Substitute the values and calculate.
$$P(\text{Group 1} | \text{first prize}) = \frac{\frac{8}{20} \times 0.06}{\frac{8}{20} \times 0.06 + \frac{7}{20} \times 0.10 + \frac{5}{20} \times 0.04}$$

Step 6: Simplify the calculation.
$$P(\text{Group 1} | \text{first prize}) = \frac{8 \times 0.06}{8 \times 0.06 + 7 \times 0.10 + 5 \times 0.04}$$
$$P(\text{Group 1} | \text{first prize}) = \frac{0.48}{0.48 + 0.70 + 0.20}$$
$$P(\text{Group 1} | \text{first prize}) = \frac{0.48}{1.38} = 0.35 = 35\%$$

Therefore, the chances that Group 1 will win the first prize are 35%.
