**Q1) An insurance company insured 1000 scooter drivers, 2000 car drivers, and 3000 truck drivers. The probability of an accident involving a scooter driver, car driver, and a truck is 0.01, 0.03, and 0.15 respectively. One of the insured persons meets with an accident. What is the probability that he is a scooter driver?**

*Explanation: This problem requires the application of Bayes' theorem to find the conditional probability.*

Step 1: Define the events.
- Let S = event that the driver is a scooter driver
- Let C = event that the driver is a car driver
- Let T = event that the driver is a truck driver
- Let A = event that the driver meets with an accident

Step 2: Identify the given information.
$$P(S) = \frac{1000}{6000} = \frac{1}{6}$$
$$P(C) = \frac{2000}{6000} = \frac{1}{3}$$
$$P(T) = \frac{3000}{6000} = \frac{1}{2}$$

$$P(A|S) = 0.01$$
$$P(A|C) = 0.03$$
$$P(A|T) = 0.15$$

Step 3: We need to find $P(S|A)$, which is the probability that the person is a scooter driver given that they had an accident.

Step 4: Apply Bayes' theorem.
$$P(S|A) = \frac{P(S) \times P(A|S)}{P(A)}$$

Step 5: Calculate the total probability $P(A)$.
$$P(A) = P(S) \times P(A|S) + P(C) \times P(A|C) + P(T) \times P(A|T)$$
$$P(A) = \frac{1}{6} \times 0.01 + \frac{1}{3} \times 0.03 + \frac{1}{2} \times 0.15$$
$$P(A) = \frac{0.01}{6} + \frac{0.03}{3} + \frac{0.15}{2}$$
$P(A) = \frac{1}{6} \times 0.01 + \frac{1}{3} \times 0.03 + \frac{1}{2} \times 0.15$
$P(A) = \frac{0.01}{6} + \frac{0.03}{3} + \frac{0.15}{2}$
$P(A) = \frac{0.01}{6} + 0.01 + 0.075$
$P(A) = 0.00167 + 0.01 + 0.075$
$P(A) = 0.08667$

Step 6: Calculate $P(S|A)$ using Bayes' theorem.
$$P(S|A) = \frac{P(S) \times P(A|S)}{P(A)}$$
$$P(S|A) = \frac{\frac{1}{6} \times 0.01}{0.08667}$$
$$P(S|A) = \frac{0.00167}{0.08667}$$
$$P(S|A) = 0.01923 \approx 0.0192$$

Therefore, the probability that the person who had an accident is a scooter driver is approximately 0.0192 or about 1.92%.

**Q2) You are planning a picnic today at Jells Park, but the morning is cloudy. 50% of all rainy days start off cloudy. But cloudy mornings are common (about 40% of days start cloudy) And this is usually a dry month (only 3 of 30 days tend to be rainy, or 10%). What is the chance of rain during the day?**

*Explanation: This problem requires the application of Bayes' theorem to find the probability of rain given that the morning is cloudy.*

Step 1: Define the events.
- Let R = event that the day is rainy
- Let C = event that the morning is cloudy
- We need to find P(R|C), the probability of rain given cloudy morning

Step 2: Identify the given information.
- P(C|R) = 0.5 (50% of all rainy days start off cloudy)
- P(C) = 0.4 (40% of days start cloudy)
- P(R) = 0.1 (10% of days in this month are rainy)

Step 3: Apply Bayes' theorem.
$$P(R|C) = \frac{P(R) \times P(C|R)}{P(C)}$$

Step 4: Substitute the values and calculate.
$$P(R|C) = \frac{0.1 \times 0.5}{0.4}$$
$$P(R|C) = \frac{0.05}{0.4}$$
$$P(R|C) = 0.125 = 12.5\%$$

Therefore, the chance of rain during the day, given that the morning is cloudy, is 12.5%.

Step 5: Interpretation.
- Although 50% of rainy days start cloudy, cloudy mornings are common (40% of all days).
- The base rate of rainy days is low (only 10%).
- This results in a relatively low probability (12.5%) that a cloudy morning will develop into a rainy day.

This example illustrates how Bayes' theorem allows us to update our probability estimates based on new evidence (the cloudy morning) and correctly account for the base rates (the frequency of cloudy mornings and rainy days in general).

**Q3) A sample of five batteries is drawn with replacement from a lot which is 5% defective. What is the probability that the number of defective batteries in the sample is exactly 2? Note: For this problem, associate 'success' with drawing a defective battery.**

*Explanation: This is a binomial probability problem where we need to find P(X = 2) with n = 5 trials and p = 0.05 probability of success (defective battery).*

Step 1: Identify that this follows a binomial distribution.
- We have a fixed number of trials (n = 5)
- Each trial has two possible outcomes (defective or non-defective)
- The trials are independent (drawing with replacement)
- The probability of success (defective battery) is constant (p = 0.05)

Step 2: Recall the binomial probability formula.

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

$$ P(X = x) = \frac{n!}{x!(n-x)!} p^x (1-p)^{n-x} $$

Where:
- $\binom{n}{k}$ is the binomial coefficient "n choose k"
- p is the probability of success
- (1-p) is the probability of failure

Step 3: Substitute the values into the formula.
$$P(X = 2) = \binom{5}{2} (0.05)^2 (0.95)^{5-2}$$

Step 4: Calculate the binomial coefficient.
$$\binom{5}{2} = \frac{5!}{2!(5-2)!} = \frac{5!}{2!3!} = \frac{5 \times 4}{2 \times 1} = 10$$

Step 5: Calculate the probability.
$$P(X = 2) = 10 \times (0.05)^2 \times (0.95)^3$$
$$P(X = 2) = 10 \times 0.0025 \times 0.857375$$
$$P(X = 2) = 10 \times 0.00214344$$
$$P(X = 2) = 0.0214344 \approx 0.0214$$

This confirms that the probability is approximately 0.0214 or 2.14%

**Q4) Imagine you're not feeling well and suspect it could be Covid-19. You then go and do a RAT test which unfortunately comes back with a positive result. Considering the test has a 95% accuracy rate — What's the probability of having Covid-19, given that it tested positive? Let's say 2 out of 100 in the population have Covid-19.**

*Explanation: This is a classic application of Bayes' theorem involving medical testing, where we need to find the probability of having the disease given a positive test result.*

Step 1: Define the events and identify the given information.
- Let D = event that a person has Covid-19
- Let T+ = event that the test result is positive
- We need to find P(D|T+), the probability of having Covid-19 given a positive test
- Given information:
  - The test has 95% accuracy rate
  - 2% of the population has Covid-19

Step 2: Clarify what "95% accuracy" means in this context.
- Sensitivity (true positive rate): P(T+|D) = 0.95 (probability of testing positive if you have Covid-19)
- Specificity (true negative rate): P(T-|~D) = 0.95 (probability of testing negative if you don't have Covid-19)
- This means the false negative rate is: P(T-|D) = 0.05
- And the false positive rate is: P(T+|~D) = 0.05

Step 3: Identify the prior probabilities.
- P(D) = 0.02 (2% of population has Covid-19)
- P(~D) = 0.98 (98% of population does not have Covid-19)

Step 4: Apply Bayes' theorem.
$$P(D|T+) = \frac{P(D) \times P(T+|D)}{P(T+)}$$

Step 5: Find P(T+), the total probability of testing positive.
$$P(T+) = P(D) \times P(T+|D) + P(~D) \times P(T+|~D)$$
$$P(T+) = 0.02 \times 0.95 + 0.98 \times 0.05$$
$$P(T+) = 0.019 + 0.049$$
$$P(T+) = 0.068$$

Step 6: Calculate P(D|T+) using Bayes' theorem.
$$P(D|T+) = \frac{P(D) \times P(T+|D)}{P(T+)}$$
$$P(D|T+) = \frac{0.02 \times 0.95}{0.068}$$
$$P(D|T+) = \frac{0.019}{0.068}$$
$$P(D|T+) = 0.2794 \approx 0.28 = 28\%$$

Therefore, despite the test being positive, the probability that you actually have Covid-19 is only about 28.6%.

Step 7: Interpretation of the result.
- This result may seem counterintuitive (why isn't it closer to 95%?), but it demonstrates the importance of considering the base rate (prior probability) in these calculations.
- When the disease is relatively rare in the population (only 2% prevalence), even a small false positive rate (5%) can lead to many false positives relative to true positives.
- This example illustrates why medical professionals need to consider the prevalence of a condition when interpreting test results, especially for screening tests in low-prevalence populations.

**Q5) Suppose X has the following probability mass function.**
- **p(0) = 0.2**
- **p(1) = 0.5**
- **p(2) = 0.3**

**Calculate E[X²].**


*Explanation: This problem involves calculating the expected value of a random variable squared.*

Step 1: Understand what we're calculating.
- The expected value of X² is: $E[X^2] = \sum x^2 \times p(x)$
- We need to calculate the sum of each possible value of X squared, multiplied by its probability

Step 2: Identify the probability mass function.
- X can take values 0, 1, and 2
- p(0) = 0.2 (probability that X = 0)
- p(1) = 0.5 (probability that X = 1)
- p(2) = 0.3 (probability that X = 2)

Step 3: Verify that the probabilities sum to 1.
$$p(0) + p(1) + p(2) = 0.2 + 0.5 + 0.3 = 1.0$$

Step 4: Calculate E[X²].
$$E[X^2] = \sum x^2 \times p(x)$$
$$E[X^2] = 0^2 \times p(0) + 1^2 \times p(1) + 2^2 \times p(2)$$
$$E[X^2] = 0^2 \times 0.2 + 1^2 \times 0.5 + 2^2 \times 0.3$$
$$E[X^2] = 0 \times 0.2 + 1 \times 0.5 + 4 \times 0.3$$
$$E[X^2] = 0 + 0.5 + 1.2$$
$$E[X^2] = 1.7$$

Therefore, E[X²] = 1.7.

For reference, we can also calculate E[X]:
$$E[X] = \sum x \times p(x) = 0 \times 0.2 + 1 \times 0.5 + 2 \times 0.3 = 0 + 0.5 + 0.6 = 1.1$$

Note that E[X²] ≠ (E[X])² in general. Here, (E[X])² = 1.1² = 1.21, which is different from E[X²] = 1.7.

**Q6) The time, in hours, it takes to locate and repair an electrical breakdown in a certain factory is a random variable — call it X — whose density function is given by:**

$$f_X(x) = \begin{cases}
1 & \text{if } 0 < x < 1 \\
0 & \text{otherwise}
\end{cases}$$

**If the cost involved in a breakdown of duration x is x³, what is the expected cost of such a breakdown?**

*Explanation: This problem involves finding the expected value of a function of a continuous random variable.*

Step 1: Identify what we're calculating.
- We have a random variable X representing repair time
- The cost function is given as C(x) = x³
- We need to find the expected cost E[C(X)] = E[X³]

Step 2: Confirm that the probability density function (PDF) is valid.
- For a PDF to be valid, it must integrate to 1 over its domain
- Verify: $\int_{0}^{1} 1 \, dx = x \big|_{0}^{1} = 1 - 0 = 1$ 
- This means X follows a uniform distribution on the interval (0,1)

Step 3: Calculate the expected value of the cost function.
- For a continuous random variable, the expected value of a function g(X) is:
$$E[g(X)] = \int_{-\infty}^{\infty} g(x) \cdot f_X(x) \, dx$$

- In our case, g(X) = X³ and the PDF is non-zero only from 0 to 1:
$$E[X^3] = \int_{0}^{1} x^3 \cdot 1 \, dx = \int_{0}^{1} x^3 \, dx$$

Step 4: Compute the integral.
$$E[X^3] = \int_{0}^{1} x^3 \, dx = \frac{x^4}{4} \bigg|_{0}^{1} = \frac{1^4}{4} - \frac{0^4}{4} = \frac{1}{4} = 0.25$$

Therefore, the expected cost of a breakdown is 0.25.

Step 5: Interpretation.
- Although the repair time is uniformly distributed between 0 and 1 hours, the expected cost is 0.25 rather than some value near 0.5.
- This is because the cost grows as the cube of time, so longer repairs contribute disproportionately more to the expected cost.
- Note that for a uniform distribution on (0,1), E[X] = 0.5, but E[X³] = 0.25.

**Q7) A construction firm has recently sent in bids for 3 jobs worth (in profits) 10, 20, and 40 (thousand) dollars. If its probabilities of winning the jobs are respectively 0.2, 0.8, and 0.3, what is the firm's expected total profit?**

*Explanation: This problem requires calculating the expected value of the total profit using the linearity of expectation.*

Step 1: Identify the random variables and their distributions.
- Let X₁, X₂, and X₃ be binary random variables representing whether the firm wins each job
- X₁ = 1 with probability 0.2 (first job - $10,000)
- X₂ = 1 with probability 0.8 (second job - $20,000)
- X₃ = 1 with probability 0.3 (third job - $40,000)

Step 2: Define the profit function.
- The profit from each job is either the full amount (if won) or 0 (if lost)
- Total profit = 10X₁ + 20X₂ + 40X₃ thousand dollars

Step 3: Calculate the expected profit using linearity of expectation.
- The expected value of a sum equals the sum of expected values
- E[Total Profit] = E[10X₁ + 20X₂ + 40X₃]
- E[Total Profit] = 10E[X₁] + 20E[X₂] + 40E[X₃]

Step 4: Calculate the expected value of each binary random variable.
- E[X₁] = 1 × P(X₁=1) + 0 × P(X₁=0) = 1 × 0.2 + 0 × 0.8 = 0.2
- E[X₂] = 1 × P(X₂=1) + 0 × P(X₂=0) = 1 × 0.8 + 0 × 0.2 = 0.8
- E[X₃] = 1 × P(X₃=1) + 0 × P(X₃=0) = 1 × 0.3 + 0 × 0.7 = 0.3

Step 5: Substitute these values to find the expected total profit.
$$E[\text{Total Profit}] = 10 \times 0.2 + 20 \times 0.8 + 40 \times 0.3$$
$$E[\text{Total Profit}] = 2 + 16 + 12$$
$$E[\text{Total Profit}] = 30 \text{ thousand dollars}$$

That's $30,000 in total expected profit.

**Q8) Compute Var(X) when X represents the outcome when we roll a fair die. Assume that it is a 6-sided die.**

*Explanation: This problem involves calculating the variance of a discrete uniform random variable representing a fair die roll.*

Step 1: Identify the random variable and its distribution.
- X represents the outcome of rolling a fair 6-sided die
- X can take values 1, 2, 3, 4, 5, 6 with equal probability
- P(X = i) = 1/6 for i = 1, 2, 3, 4, 5, 6

Step 2: Recall the formula for variance.
$$\text{Var}(X) = E[X^2] - (E[X])^2$$

Step 3: Calculate the expected value E[X].
$$E[X] = \sum_{i=1}^{6} i \times P(X = i) = \sum_{i=1}^{6} i \times \frac{1}{6}$$
$$E[X] = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = \frac{21}{6} = 3.5$$

Step 4: Calculate the expected value of X².
$$E[X^2] = \sum_{i=1}^{6} i^2 \times P(X = i) = \sum_{i=1}^{6} i^2 \times \frac{1}{6}$$
$$E[X^2] = \frac{1^2 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2}{6} = \frac{1 + 4 + 9 + 16 + 25 + 36}{6} = \frac{91}{6} = 15.17$$

Step 5: Calculate the variance.
$$\text{Var}(X) = E[X^2] - (E[X])^2 = 15.17 - (3.5)^2 = 15.17 - 12.25 = 2.92$$

Therefore, Var(X) = 2.92.

Step 6: Alternative calculation for verification.
A known formula for the variance of a discrete uniform distribution on integers from a to b is:
$$\text{Var}(X) = \frac{(b-a+1)^2 - 1}{12}$$

For a fair 6-sided die (a=1, b=6):
$$\text{Var}(X) = \frac{(6-1+1)^2 - 1}{12} = \frac{36 - 1}{12} = \frac{35}{12} = 2.92$$

This confirms our earlier calculation that Var(X) = 2.92.
